 # **<font color="DarkBlue">Generando un gr치fico de dispersi칩n</font>**

<p align="justify">
Un <b>gr치fico de dispersi칩n</b> es una visualizaci칩n que representa la relaci칩n entre dos variables num칠ricas. En este gr치fico, cada punto corresponde a una observaci칩n en el conjunto de datos, y su posici칩n en los ejes X e Y depende de sus valores en ambas variables. Este tipo de gr치fico es ideal para identificar patrones, tendencias o relaciones entre las variables, lo que puede ayudar a visualizar correlaciones, detectar outliers o entender c칩mo var칤a una variable en funci칩n de la otra.
<br><br>
En an치lisis de datos y en el contexto de negocios, el gr치fico de dispersi칩n es muy 칰til para explorar hip칩tesis o realizar an치lisis exploratorios. Por ejemplo, puede ayudar a analizar la relaci칩n entre el presupuesto de marketing y las ventas generadas. Un patr칩n creciente podr칤a indicar que a mayor inversi칩n en marketing, mayores son las ventas. Sin embargo, tambi칠n se puede observar si existe un punto de saturaci칩n donde el incremento en el gasto ya no se traduce en mayores ventas, lo que ser칤a valioso para optimizar el presupuesto.
</p>


<p align="justify"> 游 Ejemplo:  
<br><br>
Gr치fico de dispersi칩n para analizar la relaci칩n entre el presupuesto de marketing y las ventas en una empresa.

```python
# Importar las librer칤as necesarias
import pandas as pd
import plotly.express as px
```

```python
# Crear un DataFrame de ejemplo con datos de negocios
data = {'Presupuesto_Marketing': [500, 700, 1200, 1500, 2000, 2500, 3000, 3500, 4000, 4500],
        'Ventas': [10000, 15000, 20000, 21000, 25000, 27000, 29000, 31000, 32000, 32500]}

df = pd.DataFrame(data)
df
```

```python
px.scatter(df, 'Presupuesto_Marketing', 'Ventas')
```

```python
# Crear el gr치fico de dispersi칩n
px.scatter(df,
           x='Presupuesto_Marketing',
           y='Ventas',
           title='Relaci칩n entre Presupuesto de Marketing y Ventas',
           labels={'Presupuesto_Marketing':'Presupuesto de Marketing ($)', 'Ventas':'Ventas ($)'},
           trendline="ols")
```

<p align="justify">
<br>
游늵 En este caso, el gr치fico de dispersi칩n muestra cada punto como una combinaci칩n de presupuesto de marketing (eje X) y ventas generadas (eje Y).
<br><br>
La tendencia es agregada autom치ticamente mediante una l칤nea de regresi칩n (trendline="ols"), lo que permite ver c칩mo las ventas tienden a aumentar con el presupuesto de marketing. Esto da una visi칩n clara de si la inversi칩n en marketing est치 siendo efectiva y si podr칤a estar llegando a un punto de saturaci칩n donde el retorno en ventas comienza a estabilizarse.

<p align="justify">
Para mejorar el gr치fico, podemos aplicar algunas personalizaciones adicionales.
<br><br>
A continuaci칩n, algunos ajustes que pueden enriquecer la visualizaci칩n:
<br><br>

- **Modificar el Tama침o de los Puntos**: Aumentar el tama침o de los puntos para mejorar la visibilidad.
- **Ajustar los Colores**: Utilizar un esquema de color que resalte los puntos, como un degradado o un color espec칤fico para que sea visualmente atractivo.
- **Agregar Anotaciones**: Incluir etiquetas en algunos puntos clave para destacar observaciones relevantes, como el presupuesto m치s bajo o el m치s alto.
- **Personalizar la L칤nea de Tendencia**: Ajustar el estilo de la l칤nea de tendencia para que sea m치s visible, por ejemplo, cambiando su grosor o color.

```python
# Crear el gr치fico de dispersi칩n mejorado
fig = px.scatter(df,
                 x='Presupuesto_Marketing',
                 y='Ventas',
                 title='Relaci칩n entre Presupuesto de Marketing y Ventas',
                 labels={'Presupuesto_Marketing':'Presupuesto de Marketing ($)', 'Ventas':'Ventas ($)'},
                 trendline="ols",
                 size_max=10,                           # Tama침o m치ximo de los puntos
                 color='Ventas',                        # Degradado de color en funci칩n de las ventas
                 color_continuous_scale="Viridis")      # Escala de color


# Personalizar la l칤nea de tendencia
fig.update_traces(marker=dict(size=10),                   # Ajustar tama침o de puntos
                  line=dict(color="royalblue", width=2))  # Color y grosor de l칤nea de tendencia



# Agregar anotaciones en puntos clave
fig.add_annotation(x=500, y=10000, text="Inicio de inversi칩n", showarrow=True, arrowhead=5)


# Mejorar el dise침o visual
fig.update_layout(title_font_size=20,
                  xaxis_title_font_size=14,
                  yaxis_title_font_size=14,
                  template="gridon")  # Fondo blanco para mayor claridad



# Mostrar el gr치fico
fig.show()
```

- **Tama침o y Color de los Puntos**: Se ha ajustado el tama침o y color de los puntos para que reflejen visualmente los niveles de ventas y sean m치s visibles.
- **L칤nea de Tendencia**: Se ha personalizado el color y grosor de la l칤nea de tendencia para que sea m치s destacada.
- **Anotaciones**: Se han a침adido etiquetas para marcar puntos relevantes en el gr치fico, lo que ayuda a contextualizar el an치lisis.
- **Dise침o General**: Usar un fondo blanco y una paleta de colores con alto contraste facilita la legibilidad del gr치fico.

 # **<font color="DarkBlue">Generando un gr치fico de l칤nea</font>**

<p align="justify">
Un <b>gr치fico de l칤nea</b> es una visualizaci칩n que muestra c칩mo cambia una variable a lo largo de otra variable, t칤picamente el tiempo. Este gr치fico es ideal para identificar tendencias y patrones temporales en una serie de datos, ya que permite observar la direcci칩n general del cambio: si una variable aumenta, disminuye o se mantiene estable.
<br><br>
En el contexto de an치lisis de negocios, el gr치fico de l칤nea es ampliamente utilizado para visualizar indicadores de rendimiento, como ventas mensuales, ingresos anuales o el precio de las acciones de una empresa. Estos gr치ficos ofrecen una visi칩n clara de c칩mo evoluciona una variable en el tiempo, permitiendo identificar estacionalidades, patrones c칤clicos o posibles anomal칤as en el comportamiento de los datos, factores esenciales para tomar decisiones estrat칠gicas.</p>


<p align="justify"> 游 Ejemplo:  
<br><br>
Gr치fico de l칤nea para analizar las ventas mensuales de una empresa a lo largo del a침o.

```python
# Crear un DataFrame de ejemplo con datos de ventas mensuales
data = {'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
        'Ventas': [12000, 15000, 13000, 16000, 17000, 15000, 18000, 20000, 19000, 21000, 23000, 22000]}

df = pd.DataFrame(data)
df
```

```python
px.line(df, "Mes", "Ventas")
```

```python
# Crear el gr치fico de l칤nea
fig = px.line(df,
              x='Mes',
              y='Ventas',
              title='Ventas Mensuales de la Empresa',
              labels={'Ventas':'Ventas ($)'},
              markers=True)  # Agregar marcadores en cada punto

fig.show()
```

```python
# Mejorar el dise침o del gr치fico
fig.update_traces(line=dict(color="darkblue", width=2))  # Personalizar color y grosor de la l칤nea
fig.update_layout(title_font_size=18,
                  xaxis_title_font_size=14,
                  yaxis_title_font_size=14,
                  template="simple_white")               # Fondo blanco para mayor claridad


# Mostrar el gr치fico
fig.show()
```

<p align="justify">
Este gr치fico de l칤nea muestra la evoluci칩n de las ventas mensuales en el transcurso del a침o.
<br><br>
Los puntos de datos est치n marcados para destacar el valor de ventas en cada mes, y la l칤nea conecta estos puntos para reflejar la tendencia general.
<br><br>
En el ejemplo, se puede observar si las ventas aumentan en ciertos meses (quiz치s debido a promociones estacionales o eventos) y si existen meses con bajas, permitiendo al equipo de negocios identificar oportunidades para ajustar sus estrategias comerciales.








<p align="justify">
Para mejorar el gr치fico de l칤nea y hacerlo m치s informativo, se pueden aplicar varias personalizaciones adicionales:
<br><br>

- **Destacar los Puntos Clave**: A침adir anotaciones en meses espec칤ficos que muestren picos o ca칤das notables en las ventas.
- **Cambiar el Esquema de Colores**: Usar un degradado o colores adicionales para enfatizar la evoluci칩n de las ventas.
- **Establecer L칤neas de Referencia**: Agregar una l칤nea de referencia para el promedio de ventas anuales o para un objetivo espec칤fico.
- **Ajustar el Eje X**: Rotar las etiquetas del eje X para que los nombres de los meses sean m치s legibles.


```python
# Calcular el promedio de ventas para a침adir una l칤nea de referencia

promedio_ventas = df['Ventas'].mean()
```

```python
# Crear el gr치fico de l칤nea mejorado
fig = px.line(df,
              x='Mes',
              y='Ventas',
              title='Evoluci칩n de Ventas Mensuales de la Empresa',
              labels={'Ventas':'Ventas ($)'},
              markers=True,                         # Agregar marcadores en cada punto
              hover_name='Mes'                      # Agregar nombre del Mes en la etiqueta
              )


# Personalizar la l칤nea de ventas
fig.update_traces(line=dict(color="blue", width=2))  # Color y grosor de la l칤nea principal


# Agregar l칤nea de referencia del promedio de ventas
fig.add_hline(y=promedio_ventas,
              line_dash="dash",
              line_color="green",
              annotation_text="Promedio Anual de Ventas",
              annotation_position="bottom left")

# Agregar anotaciones en puntos clave (pico de ventas y mes m치s bajo)
fig.add_annotation(x="Agosto", y=20000, text="Promoci칩n de verano", showarrow=True, arrowhead=2)
fig.add_annotation(x="Junio", y=15000, text="Baja", showarrow=True, arrowhead=2)

# Mejorar la visualizaci칩n del gr치fico
fig.update_layout(
    title_font_size=18,
    xaxis_title_font_size=16,
    yaxis_title_font_size=16,
    template="plotly_white",   # Fondo blanco para claridad
    #xaxis_tickangle=-45,      # Rotaci칩n de etiquetas de eje X
)

# Mostrar el gr치fico
fig.show()
```

- **L칤nea de Promedio**: Agregar una l칤nea de referencia para el promedio anual de ventas ayuda a ver f치cilmente c칩mo se comparan los valores mensuales con el rendimiento promedio.
- **Anotaciones en Puntos Clave**: Se han incluido anotaciones para se침alar un aumento en agosto (probablemente por una promoci칩n de verano) y una baja estacional en junio, proporcionando contexto sobre las fluctuaciones.
- **Ajustes de Visualizaci칩n**: La rotaci칩n de las etiquetas de los meses facilita la lectura, y el ajuste en el grosor y color de la l칤nea ayuda a destacar el gr치fico y hacerlo visualmente m치s atractivo.







 # **<font color="DarkBlue">Generando un gr치fico de barras</font>**

<p align="justify">
Un <b>gr치fico de barras</b> es una visualizaci칩n que muestra datos categ칩ricos representados mediante barras rectangulares. La longitud o altura de cada barra es proporcional al valor de la categor칤a a la que representa.

<br><br>
En el contexto de negocios, el gr치fico de barras es 칰til para visualizar datos como ventas por regi칩n, ingresos por producto o el rendimiento de cada departamento en una organizaci칩n. <p align="justify"> 游 Ejemplo:  
<br><br>
Gr치fico de barras para analizar las ventas de diferentes productos en una empresa.

```python
# Crear un DataFrame de ejemplo con datos de ventas de productos
data = {'Producto': ['Producto A', 'Producto B', 'Producto C', 'Producto D', 'Producto E'],
        'Ventas': [25000, 30000, 15000, 20000, 27000]
}

df = pd.DataFrame(data)
df
```

<p align="justify">
En este caso, el gr치fico de barras permite comparar las ventas de cinco productos diferentes. Cada barra muestra el valor de ventas correspondiente, con un esquema de color que resalta visualmente los valores m치s altos y m치s bajos.
<br><br>
Este gr치fico facilita la identificaci칩n de los productos con mejor rendimiento (por ejemplo, el "Producto B") y aquellos con ventas m치s bajas, lo cual puede servir como base para ajustar estrategias comerciales o de marketing.

Ahora el color de cada barra, sera en funci칩n de los productos.

 # **<font color="DarkBlue">Generando un gr치fico de torta o circular</font>**

 # **<font color="DarkBlue">Generando un gr치fico de burbujas</font>**

<p align="justify">
Un <b>gr치fico de burbujas</b> es una extensi칩n de un gr치fico de dispersi칩n en el cual los puntos son reemplazados por burbujas de tama침o variable.
<br><br>
Este gr치fico permite representar tres dimensiones de datos en un solo gr치fico: las posiciones de las burbujas en los ejes X y Y indican dos variables, mientras que el tama침o de cada burbuja representa una tercera variable cuantitativa. Es especialmente 칰til para comparar categor칤as y observar relaciones entre m칰ltiples variables.
<br><br>
En el an치lisis de negocios, el gr치fico de burbujas es 칰til para representar relaciones complejas, como la correlaci칩n entre el costo, el rendimiento de inversi칩n y el tama침o del mercado para una cartera de productos, o para analizar el impacto de tres factores en la rentabilidad de cada sucursal de una empresa.
<br><br>
Este tipo de gr치fico facilita la comparaci칩n visual y ayuda a identificar r치pidamente patrones o anomal칤as, lo que puede influir en las decisiones estrat칠gicas.</p>


<p align="justify"> 游 Ejemplo:  
<br><br>
Gr치fico de burbujas para analizar la relaci칩n entre las ventas, el gasto en marketing y el n칰mero de empleados en diferentes sucursales de una empresa.

```python
# Crear un DataFrame de ejemplo con datos de sucursales
data = {'Sucursal': ['Sucursal A', 'Sucursal B', 'Sucursal C', 'Sucursal D', 'Sucursal E'],
        'Ventas': [40000, 55000, 30000, 48000, 62000],  # Ventas en d칩lares
        'Gasto_Marketing': [10000, 15000, 7000, 12000, 18000],  # Gasto en marketing en d칩lares
        'Empleados': [50, 80, 40, 60, 90]  # N칰mero de empleados en cada sucursal
}

df = pd.DataFrame(data)
df
```

```python
# Crear el gr치fico de burbujas
fig = px.scatter(df,
                 x='Gasto_Marketing',
                 y='Ventas',
                 size='Empleados',
                 color='Sucursal',
                 title='Relaci칩n entre Gasto en Marketing, Ventas y N칰mero de Empleados por Sucursal',
                 labels={'Gasto_Marketing': 'Gasto en Marketing ($)', 'Ventas': 'Ventas ($)'},
                 hover_name='Sucursal', size_max=60)


# Mejorar el dise침o del gr치fico
fig.update_layout(title_font_size=16,
                  xaxis_title_font_size=16,
                  yaxis_title_font_size=16,
                  template="plotly_white"  # Fondo blanco para mayor claridad
)

# Mostrar el gr치fico
fig.show()
```

En este gr치fico de burbujas:
<br><br>

- El eje X representa el gasto en marketing de cada sucursal.
- El eje Y representa las ventas obtenidas por cada sucursal.
- El tama침o de cada burbuja indica la cantidad de empleados en cada sucursal, destacando aquellas con mayor personal.

<br>
<p align="justify"> Este gr치fico facilita la identificaci칩n de relaciones entre variables clave. Por ejemplo, si una sucursal con un gasto en marketing relativamente bajo y pocos empleados logra buenas ventas, podr칤a ser interesante investigar su estrategia.
