 # **<font color="DarkBlue">Definición simplificada de ETL</font>**

<p align="justify">
Leer datos y hacerlos accesibles, a veces denominado carga de datos, uno de los pasos del <b>ETL</b>:
<ul><li>Extract (Extracción)</li>
<li>Transform (Transformación)</li>
<li>Load (Carga)</li></ul></p> <p align="justify">
Este es el primer paso esencial para cualquier tipo de análisis o procesamiento de datos. En ocasiones, el término "análisis" se utiliza para describir la carga e interpretación de datos en formato texto, transformándolos en tablas y diferentes tipos de datos estructurados.
<br><br>
En este módulo, nos enfocaremos en la entrada y salida de datos utilizando la biblioteca <code>Pandas</code>, aunque existen muchas otras herramientas en diferentes bibliotecas que permiten realizar las mismas tareas.
</p>

 # **<font color="DarkBlue">ETL versus ELT</font>**

<p align="justify">
La diferencia entre <b>ETL</b> (Extract, Transform, Load) y <b>ELT</b> (Extract, Load, Transform) radica principalmente en el orden en que se ejecutan las etapas del proceso y en el entorno donde se llevan a cabo las transformaciones de los datos:

 ## **<font color="DarkBlue">ETL (Extract, Transform, Load)</font>**


1. **Extract (Extracción):** Se extraen los datos desde diversas fuentes (bases de datos, archivos, APIs, etc.).
2. **Transform (Transformación):** Los datos se transforman en un entorno intermedio antes de ser cargados en el destino final. Este paso puede incluir la limpieza de datos, la agregación, el filtrado y la transformación en un formato más adecuado para el análisis.
3. **Load (Carga):** Los datos transformados se cargan en el sistema de almacenamiento o base de datos destino, como un data warehouse.

🚀 **Características clave de ETL:**
- La transformación se realiza antes de la carga.
- Es común en sistemas tradicionales de almacenamiento de datos, como **data warehouses**.
- A menudo se utiliza cuando los datos que se cargan deben ser procesados previamente para cumplir con un esquema específico o para optimizar el análisis.



 ## **<font color="DarkBlue">ELT (Extract, Load, Transform)</font>**


1. **Extract (Extracción):** Los datos se extraen desde las fuentes.
2. **Load (Carga):** Los datos sin procesar se cargan directamente en el sistema de almacenamiento, que suele ser un **data lake** o una base de datos moderna.
3. **Transform (Transformación):** Los datos se transforman dentro del propio sistema de almacenamiento o en el destino, utilizando la potencia de procesamiento del entorno, como un data warehouse o un motor de bases de datos.

🚀 **Características clave de ELT:**
- La transformación se realiza después de la carga.
- Es común en infraestructuras basadas en **data lakes** o en la nube, donde los sistemas pueden procesar grandes volúmenes de datos sin necesidad de transformar antes de cargar.
- A menudo se usa cuando el destino es lo suficientemente potente como para manejar grandes volúmenes de datos y realizar transformaciones a escala.



 ## **<font color="DarkBlue">Resúmen de las diferencias claves:</font>**

- En **ETL**, los datos se transforman antes de ser cargados en el destino.
- En **ELT**, los datos se cargan primero y luego se transforman en el destino.

Por lo tanto:
- **ETL** es más adecuado para sistemas que requieren datos bien estructurados antes del análisis,
- mientras que **ELT** es ideal para entornos en la nube o sistemas que pueden manejar y procesar grandes volúmenes de datos sin estructurar.

<p align="justify"><br>
👀 Ambos enfoques son utilizados según las necesidades de la organización y las características de los sistemas de almacenamiento.

 # **<font color="DarkBlue">Leer y escribir datos en formato de texto</font>**

```python
import numpy as np
import pandas as pd
```

<p align="justify">
<code>Pandas</code> presenta una serie de funciones para leer datos tabulares como un objeto <code>DataFrame</code>.</p>

<p align="justify"> 👀 Esas funciones son:
<ul>
<li><code>read_csv()</code> datos delimitados desde archivo CSV, una dirección URL u objeto similar a un archivo</li>
<li><code>read_fwf()</code> datos en formato de columna de ancho fijo sin delimitadores</li>
<li><code>read_excel()</code> datos tabulares de un archivo Excel XLS o XLSX</li>
<li><code>read_hdf()</code> archivos HDF5 escritos por <code>Pandas</code></li>
<li><code>read_html()</code> todas las tablas que se encuentran en el documento HTML</li>
<li><code>read_json()</code> datos de una cadena, archivo, URL u objeto similar a un archivo JSON (JavaScript Object Notation)</li>
<li><code>read_feather()</code> formato de archivo binario Feather</li>
<li><code>read_orc()</code> formato de archivo binario Apache ORC</li>
<li><code>read_parquet()</code> formato de archivo binario Apache Parquet</li>
<li><code>read_pickle()</code> objeto almacenado por <code>Pandas</code> usando el formato pickle de <code>Python</code></li>
<li><code>read_sas()</code> conjunto de datos SAS almacenado en formato de almacenamiento del sistema SAS</li>
<li><code>read_spss()</code> archivo de datos creado por SPSS</li>
<li><code>read_sql()</code> resultados de una consulta SQL (usando SQLAlchemy)</li>
<li><code>read_sql_table()</code> tabla SQL completa (usando SQLAlchemy); equivalente a una consulta que selecciona todo</li>
<li><code>read_stata()</code> conjunto de datos del formato de archivo Stata</li>
<li><code>read_xml()</code> tabla de datos de un archivo XML</li>
</ul>
</p>

 ## **<font color="DarkBlue">Lectura de archivos en partes</font>**

<p align="justify">
Al procesar archivos muy grandes o descubrir el conjunto correcto de argumentos para procesar correctamente un archivo grande, <code>Pandas</code> permite leer solo una pequeña parte de un archivo o iterar a través de partes más pequeñas del archivo.</p>

```python
datos="https://raw.githubusercontent.com/cristiandarioortegayubro/BDS/main/datasets/beneficios-otorgados-colaborativa-202210.xlsx-Dataset.csv"
```

```python
subsidios = pd.read_csv(datos)
subsidios.head(10)
```

```python
subsidios
```

<p align="justify"> 👀 Los puntos suspensivos $...$ indican que se han omitido las filas en el medio del <code>DataFrame</code>. Se puede leer solo una pequeña cantidad de filas (evitando leer todo el archivo), con el parámetro <code>nrows</code> indicando en él, la cantidad de filas o índices que se quieren leer.
</p>

```python
pd.read_csv(datos, nrows=4)
```

```python
subsidios1 = pd.read_csv(datos, nrows=4)
subsidios1
```

```python
subsidios1.shape
```

<p align="justify"> 👀 En el caso de la funcion <code>read_csv()</code>, los parámetros más destacados, son los siguientes:
<ul>
<li><code>path()</code> indica la ubicación del sistema de archivos, URL u objeto similar a un archivo</li>
<li><code>sep()</code> caracteres o expresión regular que se usará para dividir campos en cada fila</li>
<li><code>header()</code> número de fila para nombres de columna, debería ser <code>None</code> si no hay encabezado.</li>
<li><code>index_col()</code> número de columna o nombre de columna para usar como índice</li>
<li><code>names()</code> lista de nombres de columnas</li>
<li><code>skiprows()</code> número de filas al principio del archivo para ignorar o lista de números de fila (a partir de 0) para omitir</li>
<li><code>na_values()</code> secuencia de valores para reemplazar con NA</li>
<li><code>keep_default_na()</code> usar o no la lista de valores predeterminados NA (<code>True</code> por defecto)</li>
<li><code>parse_dates()</code> intenta analizar los datos en <code>datetime</code></li>
<li><code>converters()</code> aplicaría la función <code>f</code> a los valores indicados </li>
<li><code>dayfirst()</code> al analizar fechas ambiguas, se le aplica formato internacional</li>
<li><code>verbose()</code> muestra información de análisis diversa</li>
<li><code>encoding()</code> codificación de texto</li>
<li><code>squeeze()</code> Si los datos analizados contienen solo una columna, devuelva una <code>Serie</code></li>
<li><code>thousands()</code> separador de miles</li>
<li><code>decimal()</code> separador decimal</li>
<li><code>engine()</code> motor de conversión y análisis de CSV para usar</li>
</ul>
</p>

<p align="justify">
La función <code>read_html</code> de pandas es útil para leer tablas contenidas en archivos HTML o páginas web, pero solo funcionará si los datos que deseas están en forma de tablas HTML.
```python
import yfinance as yf
```

<p align="justify">
Podemos acceder a los datos de un activo financiero utilizando la función <code>Ticker</code>. Además, existe la opción de descargar la serie de precios de uno o varios activos utilizando la función <code>download</code>. Esta última opción generará un objeto DataFrame de Pandas, que permite manipular y analizar fácilmente los datos descargados

```python
goog = yf.Ticker('GOOG')
goog
```

<p align="justify">
👀 La función <code>Ticker</code> de la biblioteca yfinance se utiliza para crear un objeto Ticker asociado con el símbolo "GOOG", que es el símbolo de cotización de las acciones de Alphabet Inc., la empresa matriz de Google. Este objeto Ticker proporciona métodos y atributos para acceder a información financiera y datos de mercado relacionados con la acción de Google.

<p align="justify">
 Las acciones de Google cotizan en la Bolsa de Valores de Nasdaq bajo el símbolo o ticker <a href="https://finance.yahoo.com/quote/GOOG/">GOOG</a>.

<p align="justify">
El objeto Ticker cuenta con métodos como <code>history</code> para obtener el historial de precios de los activos financieros, <code>info</code> para obtener información general del activo, <code>option_chain</code> para obtener una lista de las opciones disponibles para un activo subyacente, entre otros métodos que te permiten acceder a diferentes tipos de datos financieros.

<p align="justify">
✅ El método <code>history</code> al igual de <code>dowload</code> devuelve un DataFrame de Pandas que contiene los datos históricos del activo financiero especificado, con columnas como 'Open', 'High', 'Low', 'Close', 'Volume', etc. Algunos de sus hiperparámetros más importantes son:

- `period` (str): El período de tiempo para el que se desean los datos. Puede ser '1d', '1mo' o '1y'. También se puede especificar un período personalizado utilizando el formato '2d', '5d', '1wk', '1mo', '3mo', '6mo', '1y', '2y', '5y', '10y', 'ytd' (Year to Date), o un rango específico de fechas ('yyyy-mm-dd','yyyy-mm-dd'). Por defecto el período de tiempo es 1 mes ('1mo').

- `interval` (str): El intervalo de tiempo para los datos. Puede ser '1m' para minutos, '5m' para 5 minutos, '15m' para 15 minutos, '30m' para 30 minutos, '60m' para 60 minutos, '90m' para 90 minutos, '1h' para 1 hora, '1d' para 1 día, '5d' para 5 días, '1wk' para 1 semana, o '1mo' para 1 mes. Por defecto el intervalo o periodicidad de los datos es diario.

- `start` (str o datetime): La fecha de inicio de los datos históricos. Puede ser especificada como un string en formato 'yyyy-mm-dd' o como un objeto datetime de Python. Si no se proporciona, se utilizará la fecha más temprana disponible.

- `end` (str o datetime): La fecha de fin de los datos históricos. Puede ser especificada como un string en formato 'yyyy-mm-dd' o como un objeto datetime de Python. Si no se proporciona, se utilizará la fecha más reciente disponible.

- `actions` (bool): Un booleano que indica si se deben incluir eventos como dividendos y divisiones de acciones en los datos históricos. Por defecto es True.
