```python
import pandas as pd
import plotly.express as px
```

<p align="justify">
Ahora vamos a ver un ejemplo m谩s complejo que utiliza groupby en un DataFrame de ventas, con varias columnas adicionales para hacer un an谩lisis m谩s detallado.
<br><br>
En este caso, el DataFrame incluye informaci贸n sobre los productos vendidos, la categor铆a, la regi贸n de la venta y el mes, adem谩s del monto de la venta y la cantidad vendida.



```python
# Generando un dataframe de ejemplo

ventas = pd.DataFrame({
    'Producto': ['Laptop', 'Smartphone', 'Camiseta', 'Pantalones', 'Televisor', 'Zapatillas',
                 'Laptop', 'Smartphone', 'Camiseta', 'Pantalones', 'Televisor', 'Zapatillas'],
    'Categor铆a': ['Electr贸nica', 'Electr贸nica', 'Ropa', 'Ropa', 'Electr贸nica', 'Ropa',
                  'Electr贸nica', 'Electr贸nica', 'Ropa', 'Ropa', 'Electr贸nica', 'Ropa'],
    'Regi贸n': ['Norte', 'Sur', 'Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur'],
    'Mes': ['Enero', 'Enero', 'Enero', 'Febrero', 'Febrero', 'Febrero', 'Enero', 'Febrero', 'Enero', 'Febrero', 'Enero', 'Febrero'],
    'Monto de Venta': [1200, 800, 250, 400, 1500, 600, 1300, 900, 300, 450, 1600, 650],
    'Cantidad Vendida': [3, 5, 10, 7, 2, 8, 4, 6, 9, 5, 3, 7]
})
```

```python
ventas
```

```python
# Agrupar por 'Categor铆a', 'Regi贸n' y 'Mes' y calcular el monto total de ventas y la cantidad total vendida
ventas_por_categoria_region_mes = ventas.groupby(['Categor铆a', 'Regi贸n', 'Mes']).agg(
    Ventas_Totales=('Monto de Venta', 'sum'),
    Cantidad_Total=('Cantidad Vendida', 'sum')
).reset_index()

ventas_por_categoria_region_mes
```

<p align="justify">
Este resultado muestra el monto total de ventas y la cantidad total vendida para cada combinaci贸n de categor铆a, regi贸n y mes.







```python
px.bar(ventas_por_categoria_region_mes,
       x='Mes',
       y='Ventas_Totales',
       color='Regi贸n',
       facet_col='Categor铆a',
       title='Ventas Totales por Categor铆a, Regi贸n y Mes',
       labels={'Ventas_Totales': 'Monto de Ventas',
               'Mes': 'Mes',
               'Regi贸n': 'Regi贸n'},
       template='gridon',
       barmode='stack'  # Gr谩fica apilada
)
```

```python
# Definir el orden de los meses

mes_order = ["Enero", "Febrero"]
```

```python
px.bar(ventas_por_categoria_region_mes,
       x='Mes',
       y='Ventas_Totales',
       color='Regi贸n',
       facet_col='Categor铆a',
       title='Ventas Totales por Categor铆a, Regi贸n y Mes',
       labels={'Ventas_Totales': 'Monto de Ventas', 'Mes': 'Mes', 'Regi贸n': 'Regi贸n'},
       barmode='stack',  # Gr谩fica apilada
       template='gridon',
       text='Ventas_Totales',
       category_orders={'Mes': mes_order}  # Ordenar el eje x por el orden deseado
)
```

 ## **<font color="DarkBlue">Count</font>**

<p align="justify">
Vamos a crear un ejemplo donde usamos la agregaci贸n count() para contar cu谩ntas veces se vende un producto dentro de cada categor铆a, regi贸n y mes. Esta t茅cnica es 煤til para obtener una frecuencia de ventas en lugar de sumar el monto total o la cantidad vendida.
<br><br>
En este caso, vamos a crear un DataFrame similar al anterior, pero el objetivo ser谩 contar el n煤mero de transacciones (o filas) que ocurren para cada combinaci贸n de Categor铆a, Regi贸n y Mes.





```python
# Agrupar los datos usando count() para contar transacciones

transacciones_por_categoria_region_mes = ventas.groupby(['Categor铆a', 'Regi贸n', 'Mes']).agg(
    Total_Transacciones=('Producto', 'count')
).reset_index()
```

```python
transacciones_por_categoria_region_mes
```

```python
px.bar(
    transacciones_por_categoria_region_mes,
    x='Mes',
    y='Total_Transacciones',
    color='Regi贸n',
    facet_col='Categor铆a',
    title='N煤mero de Transacciones por Categor铆a, Regi贸n y Mes',
    labels={'Total_Transacciones': 'N煤mero de Transacciones', 'Mes': 'Mes', 'Regi贸n': 'Regi贸n'},
    barmode='stack',
    template='gridon',
    text='Total_Transacciones',
    category_orders={'Mes': mes_order}
)
```

<p align="justify">
Esta visualizaci贸n muestra c贸mo var铆a la cantidad de transacciones en cada mes, por categor铆a y por regi贸n. El conteo de transacciones permite analizar la frecuencia de ventas, destacando patrones de demanda y popularidad por zona y tipo de producto.








 ## **<font color="DarkBlue">Cumsum</font>**

<p align="justify">
Para este ejemplo, usaremos la funci贸n cumsum() para calcular la suma acumulativa de las ventas en cada combinaci贸n de categor铆a, regi贸n y mes.
<br><br>
Esto nos permite ver c贸mo se acumulan las ventas a lo largo del tiempo, lo cual puede ser 煤til para identificar tendencias de crecimiento en ventas de cada producto en distintas regiones.



```python
ventas = pd.DataFrame({
    'Producto': ['Laptop', 'Smartphone', 'Camiseta', 'Pantalones', 'Televisor', 'Zapatillas',
                 'Laptop', 'Smartphone', 'Camiseta', 'Pantalones', 'Televisor', 'Zapatillas'],
    'Categor铆a': ['Electr贸nica', 'Electr贸nica', 'Ropa', 'Ropa', 'Electr贸nica', 'Ropa',
                  'Electr贸nica', 'Electr贸nica', 'Ropa', 'Ropa', 'Electr贸nica', 'Ropa'],
    'Regi贸n': ['Norte', 'Sur', 'Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur'],
    'Mes': ['Enero', 'Enero', 'Enero', 'Febrero', 'Febrero', 'Febrero', 'Enero', 'Febrero', 'Enero', 'Febrero', 'Enero', 'Febrero'],
    'Monto de Venta': [1200, 800, 250, 400, 1500, 600, 1300, 900, 300, 450, 1600, 650],
    'Cantidad Vendida': [3, 5, 10, 7, 2, 8, 4, 6, 9, 5, 3, 7]
})
```

```python
# Ordenar por Categor铆a, Regi贸n y Mes

ventas = ventas.sort_values(by=['Categor铆a', 'Regi贸n', 'Mes'])
```

```python
# Calcular la suma acumulativa de 'Monto de Venta' por Categor铆a y Regi贸n

ventas['Venta_Acum'] = ventas.groupby(['Categor铆a', 'Regi贸n'])['Monto de Venta'].cumsum()
ventas
```

```python
px.bar(
    ventas,
    x='Mes',
    y='Monto de Venta',
    color='Regi贸n',
    facet_col='Categor铆a',
    title='Suma Acumulativa de Ventas por Categor铆a y Regi贸n',
    labels={'Venta_Acum': 'Monto de Venta Acumulado', 'Mes': 'Mes'},
    category_orders={'Mes': ['Enero', 'Febrero']},
    template='gridon',
    text='Venta_Acum',
    barmode='stack'  # Modo apilado para sumar acumulativamente las regiones dentro de cada mes
)
```

 ## **<font color="DarkBlue">Mean</font>**

<p aling="justify">
Para este ejemplo, usaremos la funci贸n mean() para calcular el promedio de las ventas (Monto de Venta) para cada combinaci贸n de categor铆a, regi贸n y mes.
<br><br>
Este enfoque permite observar las ventas promedio de productos en distintas regiones y categor铆as, lo cual puede ser 煤til para identificar diferencias en el valor de las ventas entre regiones.



```python
ventas = pd.DataFrame({
    'Producto': ['Laptop', 'Smartphone', 'Camiseta', 'Pantalones', 'Televisor', 'Zapatillas',
                 'Laptop', 'Smartphone', 'Camiseta', 'Pantalones', 'Televisor', 'Zapatillas'],
    'Categor铆a': ['Electr贸nica', 'Electr贸nica', 'Ropa', 'Ropa', 'Electr贸nica', 'Ropa',
                  'Electr贸nica', 'Electr贸nica', 'Ropa', 'Ropa', 'Electr贸nica', 'Ropa'],
    'Regi贸n': ['Norte', 'Sur', 'Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur'],
    'Mes': ['Enero', 'Enero', 'Enero', 'Febrero', 'Febrero', 'Febrero', 'Enero', 'Febrero', 'Enero', 'Febrero', 'Enero', 'Febrero'],
    'Monto de Venta': [1200, 800, 250, 400, 1500, 600, 1300, 900, 300, 450, 1600, 650],
    'Cantidad Vendida': [3, 5, 10, 7, 2, 8, 4, 6, 9, 5, 3, 7]
})

ventas
```

```python
# Calcular el promedio de 'Monto de Venta' por Categor铆a, Regi贸n y Mes

ventas_promedio = ventas.groupby(['Categor铆a', 'Regi贸n', 'Mes'])['Monto de Venta'].mean().reset_index()
ventas_promedio['Monto de Venta'] = ventas_promedio['Monto de Venta'].round(2)
ventas_promedio
```

```python
px.bar(
    ventas_promedio,
    x='Mes',
    y='Monto de Venta',
    color='Regi贸n',
    facet_col='Categor铆a',
    title='Promedio de Ventas por Categor铆a, Regi贸n y Mes',
    template='gridon',
    text='Monto de Venta',
    labels={'Monto de Venta': 'Monto de Venta Promedio', 'Mes': 'Mes'},
    category_orders={'Mes': ['Enero', 'Febrero']}
)
```

<p align="justify">
Este gr谩fico permite visualizar el promedio de ventas en cada regi贸n y categor铆a por mes, facilitando la identificaci贸n de regiones con promedios de ventas m谩s altos o bajos en cada categor铆a y mes.








 ## **<font color="DarkBlue">Rank</font>**

<p align="justify">
Usaremos la funci贸n rank() para asignar un ranking a los valores de las ventas en cada combinaci贸n de categor铆a, regi贸n y mes.
<br><br>
Esto nos ayudar谩 a ver c贸mo se clasifican las ventas dentro de cada grupo, permitiendo identificar los productos, regiones o meses con el mejor desempe帽o seg煤n el valor de las ventas.



```python
# Calcular el ranking de 'Monto de Venta' por Categor铆a, Regi贸n y Mes

ventas['Rank_Ventas'] = ventas.groupby(['Categor铆a', 'Regi贸n', 'Mes'])['Monto de Venta'].rank(method='min', ascending=False)
ventas
```
