 # **<font color="DarkBlue">Datos financieros</font>**

```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
```

<p align="justify">
üè∑ <code>import yfinance as yf</code>: Esta l√≠nea importa la biblioteca <code>yfinance</code> y la asigna al alias <code>yf</code>. <code>yfinance</code> es una herramienta utilizada para descargar datos financieros hist√≥ricos y actuales directamente desde Yahoo Finance. Proporciona una forma sencilla de obtener informaci√≥n de mercado, como precios de acciones, vol√∫menes y m√°s.
</p>
<br>
<p align="justify">
üè∑ <code>import pandas as pd</code>: Esta l√≠nea importa la biblioteca <code>pandas</code> y la asigna al alias <code>pd</code>.
</p>
<br>
<p align="justify">
üè∑ <code>from datetime import datetime, timedelta</code>: Esta l√≠nea importa las clases <code>datetime</code> y <code>timedelta</code> del m√≥dulo <code>datetime</code>. <code>datetime</code> permite trabajar con fechas y horas, mientras que <code>timedelta</code> se utiliza para representar la diferencia entre dos puntos en el tiempo, facilitando el c√°lculo de intervalos de tiempo, como agregar d√≠as o semanas a una fecha.
</p>


```python
# Define la fecha de inicio y la fecha de finalizaci√≥n

start_date = '2023-12-31'
end_date = (datetime.strptime(start_date, '%Y-%m-%d') + timedelta(days=210)).strftime('%Y-%m-%d')
```

<p align="justify">
üè∑ <code>start_date = '2023-12-31'</code>: Esta l√≠nea define una variable <code>start_date</code> que almacena la fecha inicial como una cadena de texto en el formato <code>YYYY-MM-DD</code>. En este caso, la fecha de inicio es el 31 de diciembre de 2023.
</p>
<br>
<p align="justify">
üè∑ <code>end_date = (datetime.strptime(start_date, '%Y-%m-%d') + timedelta(days=210)).strftime('%Y-%m-%d')</code>:
Esta l√≠nea realiza varios pasos:
<ul>
<li><code>datetime.strptime(start_date, '%Y-%m-%d')</code>: Convierte la cadena de texto <code>start_date</code> a un objeto <code>datetime</code> utilizando el formato <code>YYYY-MM-DD</code>.</li>
<li><code>+ timedelta(days=210)</code>: Suma 210 d√≠as al objeto <code>datetime</code> generado a partir de <code>start_date</code> utilizando <code>timedelta</code>.</li>
<li><code>.strftime('%Y-%m-%d')</code>: Convierte el nuevo objeto <code>datetime</code> (con los d√≠as a√±adidos) de nuevo en una cadena de texto en el formato <code>YYYY-MM-DD</code>.</li>
</ul>
<p align="justify">
<br>
El resultado es que <code>end_date</code> contiene la fecha final, que es 210 d√≠as despu√©s de la fecha inicial (31 de diciembre de 2023).
</p>


```python
start_date, end_date
```

```python
# Descarga los datos de Mercado Libre (s√≠mbolo: MELI)

ticker = 'MELI'
data = yf.download(ticker, start=start_date, end=end_date)
```

<p align="justify">
üè∑ <code>ticker = 'MELI'</code>: Esta l√≠nea define una variable llamada <code>ticker</code> que almacena el s√≠mbolo burs√°til. En este caso, <code>'MELI'</code> se refiere a Mercado Libre, una compa√±√≠a que cotiza en bolsa. Este s√≠mbolo ser√° utilizado para descargar los datos financieros de esta acci√≥n.</p>
<br>
<p align="justify">
üè∑ <code>data = yf.download(ticker, start=start_date, end=end_date)</code>:
Esta l√≠nea utiliza el m√©todo <code>download()</code> de la biblioteca <code>yfinance</code> (importada previamente como <code>yf</code>) para descargar datos hist√≥ricos de precios de la acci√≥n de Mercado Libre (<code>MELI</code>) desde Yahoo Finance. El rango de fechas es especificado por las variables <code>start_date</code> y <code>end_date</code>, previamente definidas. La informaci√≥n descargada incluye datos como el precio de apertura, cierre, m√°ximo, m√≠nimo, volumen, etc., y se almacena en el DataFrame <code>data</code> para su posterior an√°lisis.
</p>


```python
# Muestra los primeros registros del DataFrame

data.head()
```

```python
# Dejando solo el valor de cierre

data = data['Close'].to_frame()
```

<p align="justify">
üè∑ <code>data = data['Close'].to_frame()</code>:
Esta l√≠nea selecciona la columna <code>'Close'</code> del DataFrame <code>data</code>, que contiene los precios de cierre diarios de la acci√≥n <code>MELI</code>. La columna <code>'Close'</code> se convierte en un nuevo DataFrame utilizando el m√©todo <code>to_frame()</code>. Esto es √∫til cuando quieres trabajar con una columna espec√≠fica.</p>


```python
# Reindexando

data.reset_index(inplace=True)
```

<p align="justify">
üè∑ <code>data.reset_index(inplace=True)</code>:
Esta l√≠nea restablece el √≠ndice del DataFrame <code>data</code>. Al usar <code>reset_index()</code>, el √≠ndice actual (que generalmente es la columna de fechas cuando se descargan datos financieros) se convierte en una columna normal dentro del DataFrame. El par√°metro <code>inplace=True</code> asegura que el DataFrame se modifique directamente sin necesidad de reasignarlo a otra variable. Esto es √∫til cuando se quiere trabajar con las fechas como una columna m√°s y no como √≠ndice del DataFrame, facilitando operaciones posteriores.
</p>


```python
# Agrega una columna con el nombre del d√≠a de la semana

data['Day of Week'] = data['Date'].dt.day_name()
```

<p align="justify">
üè∑ <code>data['Day of Week'] = data['Date'].dt.day_name()</code>:
Esta l√≠nea crea una nueva columna en el DataFrame <code>data</code> llamada <code>'Day of Week'</code>, que contiene el nombre del d√≠a de la semana correspondiente a cada fecha en la columna <code>'Date'</code>. Utiliza el atributo <code>dt</code> para acceder a las funcionalidades de manejo de fechas de pandas, y el m√©todo <code>day_name()</code> devuelve el nombre del d√≠a (por ejemplo, "Monday", "Tuesday", etc.). Esto es √∫til para an√°lisis de series temporales donde se desea estudiar el comportamiento de los datos seg√∫n el d√≠a de la semana.
</p>


```python
data.head()
```

```python
# Agrega una columna con el nombre del mes

data['Month Name'] = data['Date'].dt.month_name()
```

<p align="justify">
üè∑ <code>data['Month Name'] = data['Date'].dt.month_name()</code>:
Esta l√≠nea crea una nueva columna en el DataFrame <code>data</code> llamada <code>'Month Name'</code>, que contiene el nombre del mes correspondiente a cada fecha en la columna <code>'Date'</code>. Al igual que en el caso anterior, se utiliza el atributo <code>dt</code> para acceder a las funcionalidades de manejo de fechas de pandas, y el m√©todo <code>month_name()</code> devuelve el nombre del mes (por ejemplo, "January", "February", etc.). Esto es √∫til para realizar an√°lisis estacionales o entender c√≥mo se distribuyen los datos a lo largo de los diferentes meses del a√±o.
</p>


```python
data.head()
```

```python
# Redondea los valores de la columna 'Close' a dos decimales

data['Close'] = data['Close'].round(2)
```

```python
data.head()
```

```python
# Calcula la diferencia diaria en la columna 'Close'

data['Diff'] = data['Close'].diff()
```

<p align="justify">
üè∑ <code>data['Diff'] = data['Close'].diff()</code>:
Esta l√≠nea calcula la diferencia entre los precios de cierre de la acci√≥n <code>MELI</code> de un d√≠a al siguiente y almacena el resultado en una nueva columna llamada <code>'Diff'</code> dentro del DataFrame <code>data</code>. El m√©todo <code>diff()</code> de pandas devuelve la diferencia entre cada elemento y el elemento anterior en la serie. Esto es √∫til para analizar la variabilidad diaria del precio, permitiendo identificar cambios en el precio de la acci√≥n y evaluar su volatilidad a lo largo del tiempo.
</p>


```python
data.head()
```

```python
# Calcula el cambio porcentual diario en la columna 'Close'

data['% Change'] = data['Close'].pct_change() * 100
data['% Change'] = data['% Change'].round(2)
```

<p align="justify">
üè∑ <code>data['% Change'] = data['Close'].pct_change() * 100</code>:
Esta l√≠nea calcula el cambio porcentual de los precios de cierre de la acci√≥n <code>MELI</code> de un d√≠a al siguiente y lo almacena en una nueva columna llamada <code>'% Change'</code> dentro del DataFrame <code>data</code>. El m√©todo <code>pct_change()</code> de pandas calcula la variaci√≥n relativa entre el valor actual y el valor anterior, y al multiplicar por 100 se convierte en un porcentaje. Esto es √∫til para evaluar el rendimiento diario de la acci√≥n y entender las fluctuaciones en su precio.
<br><br>
üè∑ <code>data['% Change'] = data['% Change'].round(2)</code>:
Esta l√≠nea redondea los valores de la columna <code>'% Change'</code> a dos decimales. Esto ayuda a presentar los datos de forma m√°s clara y concisa, facilitando la interpretaci√≥n de los cambios porcentuales sin perder precisi√≥n significativa en la informaci√≥n.
</p>


```python
data.head()
```

```python
# Eliminamos los valores nulos

data = data.dropna()
data.head()
```
