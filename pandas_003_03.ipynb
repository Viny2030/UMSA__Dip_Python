 ## **<font color="DarkBlue">Manipulando cadenas de caracteres</font>**

<p align="justify">
Supongamos que tienes un conjunto de datos de ventas en el que hay una columna que contiene descripciones de productos en un solo campo, y deseas separarlas en columnas distintas para facilitar el an谩lisis.
<br><br>
Esto puede ser 煤til para realizar un an谩lisis m谩s detallado de los productos vendidos y mejorar la segmentaci贸n de los datos.



```python
import pandas as pd
```

 Ejemplo:

```python
# Crear un DataFrame de ejemplo
data = {
    'Ventas': [100, 200, 150, 300],
    'Productos': [
        'Laptop - Dell - 15.6 pulgadas',
        'Tel茅fono - Apple - iPhone 13',
        'Tablet - Samsung - Galaxy Tab',
        'Laptop - HP - 14 pulgadas'
    ]
}

df = pd.DataFrame(data)

print("DataFrame Original:")
df
```

```python
# Usar split() para dividir la columna 'Productos' en m煤ltiples columnas
# Usamos expand=True para crear nuevas columnas

df[['Tipo', 'Marca', 'Modelo']] = df['Productos'].str.split(' - ', expand=True)
```

```python
print("\nDataFrame Modificado:")
df
```

 ## **<font color="DarkBlue">Datos categ贸ricos</font>**

<p align="justify">
Supongamos que tenemos un DataFrame que contiene informaci贸n sobre las ventas de diferentes productos en una tienda. Queremos analizar las ventas de ciertas categor铆as de productos y utilizar el m茅todo <code>take()</code> para seleccionar filas espec铆ficas del DataFrame.

```python
# Crear un DataFrame de ejemplo con datos de ventas
data = {
    'Producto': ['Laptop', 'Smartphone', 'Tablet', 'Smartwatch', 'Laptop', 'Smartphone', 'Tablet', 'Smartwatch'],
    'Categor铆a': ['Electr贸nica', 'Electr贸nica', 'Electr贸nica', 'Accesorios', 'Electr贸nica', 'Electr贸nica', 'Electr贸nica', 'Accesorios'],
    'Ventas': [1200, 800, 300, 150, 1400, 900, 450, 200],
    'Cantidad': [5, 4, 2, 3, 7, 5, 3, 2]
}

df = pd.DataFrame(data)

print("Datos de Ventas:")
df
```

```python
# Filtrar las filas donde la categor铆a es 'Electr贸nica'

categor铆a_electr贸nica = df[df['Categor铆a'] == 'Electr贸nica']
categor铆a_electr贸nica
```

```python
# Usar el m茅todo take() para seleccionar las primeras 3 filas de la categor铆a 'Electr贸nica'

resultados_electr贸nicos = categor铆a_electr贸nica.take([0, 1, 2])
resultados_electr贸nicos
```

 ## **<font color="DarkBlue">M茅todos categ贸ricos en series</font>**

<p align="justify">
Las series que contienen datos categ贸ricos tienen varios m茅todos especiales
<br><br>
Imaginemos que una tienda de electr贸nica vende varios tipos de productos y desea analizar las ventas por categor铆a
<br><br>
 Usaremos una Serie de Pandas para representar las categor铆as de frutas y aplicaremos m茅todos categ贸ricos como:

- value_counts(),
- cat.codes, y
- cat.categories.





```python
# Crear una Serie de Pandas con datos categ贸ricos de productos vendidos

productos_vendidos = pd.Series(['Smartphone', 'Laptop', 'Tablet', 'Smartphone', 'Smartwatch',
                                 'Laptop', 'Tablet', 'Smartwatch', 'Headphones', 'Laptop'])
```

```python
# Convertir la Serie en un tipo categ贸rico

productos_categoricos = productos_vendidos.astype('category')
```

```python
# Mostrar la Serie original y su tipo

print("Serie Original de Productos Vendidos:")
productos_vendidos.name = "Productos Vendidos"
productos_vendidos
```

```python
print("\nTipo Categ贸rico de Productos:")
productos_categoricos.name = "Categ贸rico de Productos"
productos_categoricos
```

```python
# Usar value_counts() para contar las ocurrencias de cada categor铆a

conteo_productos = productos_categoricos.value_counts()
print("\nConteo de Productos Vendidos:")
conteo_productos.name = "Conteo de Productos"
conteo_productos
```

```python
# Obtener los c贸digos de las categor铆as

codigos_productos = productos_categoricos.cat.codes
print("\nC贸digos de los Productos:")
codigos_productos.name = "C贸digos de Productos"
codigos_productos
```

```python
# Obtener las categor铆as 煤nicas

categorias_unicas = productos_categoricos.cat.categories
print("\nCategor铆as nicas de Productos:")
categorias_unicas
```

```python
# Modificar las categor铆as (a帽adir una nueva categor铆a)

productos_categoricos = productos_categoricos.cat.add_categories(['Camera'])
print("\nCategor铆as Actualizadas de Productos:")
productos_categoricos.cat.categories
```

```python
# Asignar algunos productos a la nueva categor铆a

productos_categoricos[0] = 'Camera'  # Cambiando 'Smartphone' a 'Camera'
```

```python
print("\nSerie de Productos Vendidos Actualizada:")
productos_categoricos.name = "Productos Vendidos"
productos_categoricos
```

```python
# Verificar los nuevos c贸digos y conteos despu茅s de la modificaci贸n

print("\nC贸digos de los Productos Actualizados:")
productos_actualizados = productos_categoricos.cat.codes
productos_actualizados.name = "C贸digos de Productos"
productos_actualizados
```

```python
print("\nConteo de Productos Vendidos Actualizado:")
productos_categoricos.value_counts()
```

Los m茅todos categ贸ricos que podemos utilizar en las series, son los siguientes:



<ul>
    <li><strong>add_categories:</strong> A帽adir categor铆as nuevas (no utilizadas) al final de las categor铆as existentes.</li>
    <li><strong>as_ordered:</strong> Hacer categor铆as ordenadas.</li>
    <li><strong>as_unordered:</strong> Hacer que las categor铆as est茅n desordenadas.</li>
    <li><strong>remove_categories:</strong> Eliminar categor铆as, estableciendo cualquier valor eliminado como nulo.</li>
    <li><strong>remove_unused_categories:</strong> Eliminar cualquier valor de categor铆a que no aparezca en los datos.</li>
    <li><strong>rename_categories:</strong> Reemplazar categor铆as con el conjunto indicado de nuevos nombres de categor铆as; no se puede cambiar el n煤mero de categor铆as.</li>
    <li><strong>reorder_categories:</strong> Se comporta como rename_categories, pero tambi茅n puede cambiar el resultado para tener categor铆as ordenadas.</li>
    <li><strong>set_categories:</strong> Reemplace las categor铆as con el conjunto indicado de nuevas categor铆as; puede agregar o eliminar categor铆as.</li>
</ul>
