 # **<font color="DarkBlue">Objetos del indice</font>**

<p align="justify">
Los objetos de 칤ndice de un DataFrame en Pandas son una parte esencial de su estructura. El 칤ndice de un DataFrame es un conjunto de etiquetas que identifican de manera 칰nica cada fila (y cada columna, en el caso de los 칤ndices de columnas).
<br><br>
Estos 칤ndices permiten acceder, manipular, y realizar operaciones eficientes en los datos de un DataFrame.
<br><br>
Los 칤ndices pueden ser:
<br><br>

- simples, con un solo nivel (Index), o
- jer치rquicos, con m칰ltiples niveles (MultiIndex).



<p align="justify">
A continuaci칩n, algunos m칠todos y propiedades comunes asociados con los objetos de 칤ndice, junto con ejemplos pr치cticos...



 ## **<font color="DarkBlue">append()</font>**

<p align="justify">
Agrega valores adicionales al 칤ndice, creando un nuevo 칤ndice que es la concatenaci칩n de los 칤ndices originales.
<br><br>
Ejemplo: Supongamos que tienes un DataFrame con los nombres de empleados y deseas agregar m치s empleados al 칤ndice.



```python
# 칈ndice original
index_original = pd.Index(['Empleado A', 'Empleado B', 'Empleado C'])

# Agregar nuevos empleados al 칤ndice
index_nuevo = index_original.append(pd.Index(['Empleado D', 'Empleado E']))
index_nuevo
```

 ## **<font color="DarkBlue">difference()</font>**

<p align="justify">
Devuelve la diferencia entre dos 칤ndices, es decir, los elementos que est치n en un 칤ndice pero no en el otro.
<br><br>
Ejemplo: Tienes un 칤ndice de productos en stock y un 칤ndice de productos vendidos, y deseas encontrar los productos que a칰n est치n en stock.

```python
# 칈ndices de productos
index_stock = pd.Index(['Producto A', 'Producto B', 'Producto C'])
index_vendidos = pd.Index(['Producto A'])

# Encontrar productos en stock no vendidos
index_diferencia = index_stock.difference(index_vendidos)
index_diferencia
```

 ## **<font color="DarkBlue">intersection()</font>**

<p align="justify">
Devuelve la intersecci칩n de dos 칤ndices, es decir, los elementos que est치n en ambos 칤ndices.
<br><br>
Ejemplo: Tienes un 칤ndice de productos en oferta y un 칤ndice de productos con alta demanda, y deseas encontrar los productos que est치n en ambas categor칤as.

```python
# 칈ndices de productos
index_oferta = pd.Index(['Producto A', 'Producto B'])
index_demanda = pd.Index(['Producto B', 'Producto C'])

# Encontrar productos en oferta y con alta demanda
index_interseccion = index_oferta.intersection(index_demanda)
index_interseccion
```

 ## **<font color="DarkBlue">union()</font>**

<p align="justify">
Devuelve la uni칩n de dos 칤ndices, combinando los elementos de ambos 칤ndices sin duplicados.
<br><br>
Ejemplo: Combina los productos en stock y los productos que est치n en tr치nsito para obtener un listado completo de productos disponibles.

```python
# 칈ndices de productos
index_stock = pd.Index(['Producto A', 'Producto B'])
index_transito = pd.Index(['Producto C', 'Producto B'])

# Uni칩n de productos en stock y en tr치nsito
index_union = index_stock.union(index_transito)
index_union
```

 ## **<font color="DarkBlue">isin()</font>**

<p align="justify">
Devuelve un array de valores booleanos que indica si los elementos del 칤ndice est치n en otro 칤ndice o lista.
<br><br>
Ejemplo: Determina si ciertos productos est치n en la lista de productos m치s vendidos.



```python
# 칈ndice de productos
index_productos = pd.Index(['Producto A', 'Producto B', 'Producto D'])

# Lista de productos m치s vendidos
mas_vendidos = ['Producto A', 'Producto C']

# Comprobar si los productos est치n en la lista de m치s vendidos
resultados = index_productos.isin(mas_vendidos)
resultados
```

<p align="justify">

 ## **<font color="DarkBlue">delete()</font>**

<p align="justify">
Elimina un elemento en la posici칩n especificada del 칤ndice. Ejemplo: Elimina un producto del 칤ndice de productos defectuosos.



```python
# 칈ndice de productos defectuosos
index_defectuosos = pd.Index(['Producto A', 'Producto B', 'Producto C'])

# Eliminar el segundo producto del 칤ndice
index_modificado = index_defectuosos.delete(1)
index_modificado
```

 ## **<font color="DarkBlue">drop()</font>**

<p align="justify">
Elimina elementos espec칤ficos del 칤ndice. Ejemplo: Remueve un producto que ya no est치 en stock.



```python
# 칈ndice de productos en stock
index_stock = pd.Index(['Producto A', 'Producto B', 'Producto C'])

# Eliminar un producto del 칤ndice
index_modificado = index_stock.drop('Producto B')
index_modificado
```

 ## **<font color="DarkBlue">is_unique()</font>**

<p align="justify">
Una propiedad que indica si todos los elementos del 칤ndice son 칰nicos (no repetidos). Ejemplo: Verifica si los identificadores de producto en un 칤ndice son 칰nicos.



```python
# 칈ndice de identificadores de producto
index_productos = pd.Index(['ID_A', 'ID_B', 'ID_C', 'ID_A'])

# Comprobar si los identificadores son 칰nicos
es_unico = index_productos.is_unique
es_unico
```

 ## **<font color="DarkBlue">unique()</font>**

<p align="justify">
