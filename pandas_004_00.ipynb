 # **<font color="DarkBlue">Indezaci칩n jer치rquica</font>**

<p align="justify">
La indexaci칩n jer치rquica es una caracter칤stica esencial de Pandas que facilita trabajar con datos organizados en m칰ltiples niveles de 칤ndices, permitiendo tener dos o m치s niveles en un mismo eje. Esto es particularmente 칰til cuando se desea analizar datos con varias categor칤as o dimensiones, pero en un formato tabular de menor complejidad.
<br><br>
En lugar de utilizar un formato de matriz tridimensional, Pandas permite representar datos de mayor dimensi칩n en una estructura bidimensional, como un DataFrame, sin perder la capacidad de acceder y manipular datos de distintos niveles. Con la indexaci칩n jer치rquica, puedes realizar operaciones m치s complejas, como agrupar datos o seleccionar subconjuntos espec칤ficos, sin tener que reorganizar el conjunto de datos original.
</p>

游 Ejemplo:

<p align="justify">
游 Supongamos que trabajas con los datos de ventas de una empresa que tiene filiales en diferentes ciudades y vende varios productos. Quieres analizar las ventas de cada producto por ciudad, en diferentes trimestres del a침o.



```python
import pandas as pd
import numpy as np
```

```python
# Datos de ejemplo: ventas trimestrales por ciudad y producto

arrays = [
    ['Buenos Aires', 'Buenos Aires', 'Buenos Aires', 'C칩rdoba', 'C칩rdoba', 'C칩rdoba', 'Rosario', 'Rosario', 'Rosario'],
    ['Q1', 'Q2', 'Q3', 'Q1', 'Q2', 'Q3', 'Q1', 'Q2', 'Q3']
]
```

```python
arrays
```

```python
index = pd.MultiIndex.from_arrays(arrays, names=('Ciudad', 'Trimestre'))
```

```python
index
```

```python
# Ventas de productos en diferentes ciudades y trimestres

data = {'Producto A': [200, 220, 250, 180, 190, 210, 160, 170, 180],
        'Producto B': [150, 160, 170, 130, 140, 145, 120, 125, 130],
        'Producto C': [100, 110, 120, 90, 95, 105, 80, 85, 90]}
```

```python
df = pd.DataFrame(data, index=index)
```

```python
df
```

<p align="justify">
游눙 Lo que est치s viendo es una vista mejorada de un DataFrame con un MultiIndex como 칤ndice...

<p align="justify">
游낑 Entonces, con <code>MultiIndex</code> creamos un 칤ndice jer치rquico utilizando las ciudades y los trimestres como niveles del 칤ndice. Este tipo de indexaci칩n nos permite organizar los datos en diferentes niveles, de manera que podamos acceder a las ventas por ciudad y trimestre de una manera m치s ordenada.
<br><br>
Al utilizar indexaci칩n jer치rquica, puedes seleccionar datos espec칤ficos.
<br><br>
Por ejemplo, si quieres obtener las ventas del "Producto A" en "Buenos Aires" durante el primer trimestre, puedes hacer lo siguiente...



```python
# Seleccionar ventas del Producto A en Buenos Aires durante el Q1

ventas_ba_q1_PA = df.loc[('Buenos Aires', 'Q1'), 'Producto A']
ventas_ba_q1_PA
```

```python
# Seleccionar las ventas en Buenos Aires durante el Q1

ventas_ba_q1 = df.loc[('Buenos Aires', 'Q1')]
ventas_ba_q1
```

<p align="justify">
Este tipo de indexaci칩n jer치rquica es 칰til cuando manejas grandes vol칰menes de datos categorizados en varios niveles, como ventas por ubicaci칩n y periodo de tiempo. Puedes f치cilmente filtrar, agrupar o comparar las ventas por ciudad o trimestre, lo que te proporciona un an치lisis m치s flexible y detallado para la toma de decisiones empresariales.








 ## **<font color="DarkBlue">M칠todo xs</font>**

游 Vemos nuestro DataFrame

```python
df
```

游 y ahora queremos solo el segundo trimestre de todo...

```python
# Seleccionar las ventas en el segundo trimestre (Q2) para todas las ciudades y productos

ventas_q2 = df.xs('Q2', level='Trimestre')
ventas_q2
```

<p align="justify">
<code>xs()</code>: El m칠todo <code>.xs()</code> te permite seleccionar datos en un nivel espec칤fico del 칤ndice jer치rquico. Aqu칤, seleccionamos todas las ventas correspondientes al segundo trimestre (Q2) en todas las ciudades.
<br><br>
El par치metro <code>level='Trimestre'</code> indica que estamos filtrando en el nivel de trimestres. De esta forma vemos las ventas de Q2 en las ciudades de Buenos Aires, C칩rdoba y Rosario para los productos A, B y C.



```python
# Seleccionar las ventas de C칩rdoba

ventas_cordoba = df.xs('C칩rdoba', level='Ciudad')
ventas_cordoba
```

<p align="justify">
La indexaci칩n jer치rquica desempe침a un papel importante en la reestructuraci칩n de datos y en operaciones basadas en grupos, como la formaci칩n de una tabla din치mica.

 ## **<font color="DarkBlue">M칠todo unstack</font>**

```python
# Creamos un DataFrame con ventas trimestrales jer치rquicas (칤ndice jer치rquico)

index = pd.MultiIndex.from_product(
    [['Buenos Aires', 'C칩rdoba', 'Rosario'], ['Producto A', 'Producto B', 'Producto C'], ['Q1', 'Q2', 'Q3', 'Q4']],
    names=['Ciudad', 'Producto', 'Trimestre']
)
```

```python
# Generamos datos aleatorios de ventas

data = np.random.randint(100, 500, size=(36,))
df = pd.DataFrame(data, index=index, columns=['Ventas'])
```

```python
df
```

```python
# Usamos unstack() para pivotar el nivel "Trimestre" como columnas

df_unstacked = df.unstack(level='Trimestre')
print("\nDataFrame despu칠s de aplicar unstack() sobre el nivel 'Trimestre':")
df_unstacked
```

<p align="justify">
Convierte uno de los niveles del 칤ndice jer치rquico en columnas. En este caso, el nivel Trimestre pasa a ser columnas, lo que permite una visualizaci칩n m치s plana de los datos.
<br><br>
Este m칠todo es 칰til para reorganizar datos jer치rquicos. Por ejemplo, puede ser usado para analizar ventas por ciudad y producto a lo largo de varios trimestres, permitiendo comparar el rendimiento entre per칤odos de forma m치s clara cuando se deshacen los niveles de 칤ndice con <code>unstack()</code>

```python
df_unstacked
```

```python
df_unstacked.columns.names
```

 ## **<font color="DarkBlue">Reordenando</font>**

```python
df = df_unstacked
df
```

<p align="justify">
<code>swaplevel()</code> intercambia dos niveles del 칤ndice jer치rquico. En este caso, intercambiamos los niveles Ciudad y Producto, de forma que las ventas se agrupen primero por producto y luego por ciudad.

```python
# Usamos swaplevel() para intercambiar los niveles 'Ciudad' y 'Producto'

df_swapped = df.swaplevel('Ciudad', 'Producto')
print("\nDataFrame despu칠s de aplicar swaplevel('Ciudad', 'Producto'):")
df_swapped
```

<p align="justify">

<code>sort_index()</code> reorganiza el DataFrame seg칰n el 칤ndice jer치rquico, ordenando alfab칠ticamente o num칠ricamente los niveles despu칠s de haber sido intercambiados. Esto ayuda a ordenar y organizar los datos de forma m치s legible y estructurada.


```python
# Usamos sort_index() para ordenar el 칤ndice jer치rquico

df_sorted = df_swapped.sort_index()
print("\nDataFrame despu칠s de aplicar sort_index():")
df_sorted
```
