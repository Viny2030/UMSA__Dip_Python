 # **<font color="DarkBlue">Fusi贸n de datos, merge</font>**

 Ejemplo:

<p align="justify">
 Imagina que tienes dos conjuntos de datos, uno con las ventas de varios productos y otro con los datos de los empleados que realizaron las ventas. Queremos combinar esta informaci贸n para tener una visi贸n integral de qu茅 empleados realizaron qu茅 ventas.

```python
import pandas as pd
import numpy as np
```

```python
# DataFrame 1: Ventas por productos

ventas = pd.DataFrame({
    'Venta_ID': [101, 102, 103, 104],
    'Producto': ['Laptop', 'Celular', 'Tablet', 'Monitor'],
    'Monto_Venta': [1200, 800, 300, 450],
    'Empleado_ID': [1, 2, 1, 3]
})
```

<p align="justify">
DataFrame ventas: Contiene informaci贸n sobre las ventas realizadas. Cada venta tiene un Venta_ID, un producto vendido, el monto de la venta y el ID del empleado que realiz贸 la venta.

```python
ventas
```

```python
# DataFrame 2: Datos de empleados
empleados = pd.DataFrame({
    'Empleado_ID': [1, 2, 3],
    'Nombre': ['Juan', 'Mar铆a', 'Pedro'],
    'Departamento': ['Electr贸nica', 'Telefon铆a', 'Accesorios']
})
```

<p align="justify">
DataFrame empleados: Contiene los datos de los empleados, como su Empleado_ID, nombre y el departamento en el que trabajan.

```python
empleados
```

```python
# Unimos los DataFrames usando el m茅todo merge

df_merged = pd.merge(ventas, empleados, on='Empleado_ID')
```

<p align="justify">
<code>merge()</code>: El m茅todo <code>merge()</code> permite combinar estos dos DataFrames a trav茅s de una columna en com煤n, en este caso Empleado_ID. El resultado es un DataFrame que contiene toda la informaci贸n de ventas, productos y empleados.

```python
df_merged
```

<p align="justify">
El m茅todo <code>merge()</code> es 煤til en an谩lisis de datos empresariales donde los datos est谩n distribuidos en diferentes tablas.
<br><br>
Por ejemplo, puedes tener un conjunto de datos de ventas y otro conjunto de datos de empleados, y necesitar combinar esta informaci贸n para realizar un an谩lisis m谩s profundo, como identificar qu茅 empleados son responsables de las ventas m谩s altas o evaluar el desempe帽o por departamento

<p align="justify">
A continuaci贸n, se describen los comportamientos que puede adoptar el par谩metro <code>how</code> al realizar combinaciones entre tablas en pandas:
<br><br>
<strong>how="inner"</strong>: Utiliza 煤nicamente las combinaciones de teclas observadas en ambas tablas.
<br><br>
<strong>how="left"</strong>: Utiliza todas las combinaciones de teclas que se encuentran en la tabla de la izquierda.
<br><br>
<strong>how="right"</strong>: Utiliza todas las combinaciones de teclas que se encuentran en la tabla de la derecha.
<br><br>
<strong>how="outer"</strong>: Utiliza todas las combinaciones de teclas observadas en ambas tablas juntas.
</p>


<p align="justify">
Ahora vemos otro ejemplo, con dos conjuntos de datos, las ventas y los clientes, y el comportamiento de la fusi贸n de los datos lo ajustamos con el par谩metro <code>how=</code>

```python
# Creamos DataFrames

ventas = pd.DataFrame({
    'Cliente_ID': [1, 2, 3],
    'Producto': ['A', 'B', 'C'],
    'Ventas_Q1': [100, 150, 200]
})

clientes = pd.DataFrame({
    'Cliente_ID': [1, 2, 4],
    'Nombre': ['Juan', 'Mar铆a', 'Carlos']
})
```

```python
ventas
```

```python
clientes
```

 ## **<font color="DarkBlue">Inner Join</font>**

```python
# Inner Join: Combina solo las claves presentes en ambos DataFrames

inner_merge = pd.merge(ventas, clientes, on='Cliente_ID', how='inner')
print("Inner Join:")
inner_merge
```

 ## **<font color="DarkBlue">Left Join</font>**

```python
# Left Join: Utiliza todas las claves de la tabla de la izquierda (ventas)

left_merge = pd.merge(ventas, clientes, on='Cliente_ID', how='left')
print("\nLeft Join:")
left_merge
```

 ## **<font color="DarkBlue">Right Join</font>**

```python
# Right Join: Utiliza todas las claves de la tabla de la derecha (clientes)

right_merge = pd.merge(ventas, clientes, on='Cliente_ID', how='right')
print("\nRight Join:")
right_merge
```

 ## **<font color="DarkBlue">Outer Join</font>**

```python
# Outer Join: Utiliza todas las claves observadas en ambas tablas

outer_merge = pd.merge(ventas, clientes, on='Cliente_ID', how='outer')
print("\nOuter Join:")
outer_merge
```

 ## **<font color="DarkBlue">Concatenando filas</font>**

 Ejemplo:

<p align="justify">
 Imagina que tienes datos de ventas de dos sucursales diferentes de una cadena de tiendas y quieres combinar estos datos en un 煤nico DataFrame.

```python
# Datos de ventas de la sucursal 1

ventas_sucursal_1 = pd.DataFrame({
    'Producto': ['Laptop', 'Celular', 'Tablet'],
    'Monto_Venta': [1200, 800, 300],
    'Sucursal': ['Sucursal 1', 'Sucursal 1', 'Sucursal 1']
})
```

<p align="justify">
DataFrame ventas_sucursal_1: Contiene informaci贸n sobre las ventas realizadas en la sucursal 1. Para cada producto, tenemos el monto de la venta y la sucursal de origen.

```python
ventas_sucursal_1
```

```python
# Datos de ventas de la sucursal 2

ventas_sucursal_2 = pd.DataFrame({
    'Producto': ['Monitor', 'Teclado', 'Mouse'],
    'Monto_Venta': [450, 50, 25],
    'Sucursal': ['Sucursal 2', 'Sucursal 2', 'Sucursal 2']
})
```

<p align="justify">
DataFrame ventas_sucursal_2: Similar al anterior, pero con los datos de la sucursal 2.


```python
ventas_sucursal_2
```

```python
# Concatenar los datos de las dos sucursales

ventas_total = pd.concat([ventas_sucursal_1, ventas_sucursal_2])
```

<p align="justify">
El m茅todo concat() combina los dos DataFrames apil谩ndolos uno debajo del otro. En este caso, no especificamos ning煤n eje, lo que significa que concatenamos filas (por defecto)

```python
ventas_total
```

<p align="justify">
Este m茅todo es 煤til cuando tienes conjuntos de datos de diferentes fuentes o ubicaciones, como sucursales, regiones o departamentos, y necesitas combinarlos para realizar un an谩lisis conjunto. Por ejemplo, podr铆as concatenar los datos de ventas de todas tus sucursales para evaluar el rendimiento global de tu empresa o comparar las ventas entre diferentes ubicaciones.

 ## **<font color="DarkBlue">Concatenando columnas</font>**

 Ejemplo:

<p align="justify">
 Supongamos que tenemos dos DataFrames. Uno contiene informaci贸n de clientes y el otro contiene informaci贸n sobre las compras de productos que realizaron esos clientes. Queremos concatenar las columnas de estos dos DataFrames para crear una tabla completa que una la informaci贸n de clientes con los productos que han comprado.



```python
# Crear DataFrames
clientes = pd.DataFrame({
    'Cliente_ID': [1, 2, 3],
    'Nombre': ['Juan', 'Mar铆a', 'Carlos'],
    'Apellido': ['P茅rez', 'Garc铆a', 'S谩nchez']
})

compras = pd.DataFrame({
    'Cliente_ID': [1, 2, 3],
    'Producto': ['Laptop', 'Tablet', 'Smartphone'],
    'Cantidad': [1, 2, 1]
})
```

```python
clientes
```

```python
compras
```

```python
# Concatenar columnas de los dos DataFrames

data_completa = pd.concat([clientes, compras[['Producto', 'Cantidad']]], axis=1)
```

```python
data_completa
```

<p align="justify">
Este enfoque es 煤til cuando los DataFrames que se desean concatenar est谩n alineados por su 铆ndice y se quiere juntar la informaci贸n para un an谩lisis m谩s completo.
