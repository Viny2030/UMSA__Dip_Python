 ## **<font color="DarkBlue">Ejercicio 1: Gr치fico de Linea</font>**

<p><strong>Consigna:</strong> Utiliza los siguientes datos de ventas mensuales para crear un gr치fico de l칤neas que muestre la evoluci칩n de las ventas a lo largo del a침o.</p>
<ul>
    <li><strong>Meses:</strong> ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']</li>
    <li><strong>Ventas:</strong> [1200, 2300, 1800, 2500, 3000, 3200, 4100, 3800, 4500, 4900, 4800, 5200]</li>
</ul>

```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.DataFrame({"Meses": ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
"Ventas": [1200, 2300, 1800, 2500, 3000, 3200, 4100, 3800, 4500, 4900, 4800, 5200]})
```

```python
df
```

```python
fig = px.line(df,
              x="Meses",
              y="Ventas",
              title="Ventas Mensuales",
              markers=True)
fig.show()
```

```python
fig = px.line(df,
              x="Meses",
              y="Ventas",
              title="Ventas Mensuales",
              markers=True)

# Personalizar la l칤nea de tendencia
fig.update_traces(marker=dict(size=10),                   # Ajustar tama침o de puntos
                  line=dict(color="royalblue", width=2))  # Color y grosor de l칤nea de tendencia
fig.show()

```

```python
fig = px.line(df,
              x="Meses",
              y="Ventas",
              title="Ventas Mensuales",
              markers=True)

# Personalizar la l칤nea de tendencia
fig.update_traces(marker=dict(size=10),                   # Ajustar tama침o de puntos
                  line=dict(color="royalblue", width=2))  # Color y grosor de l칤nea de tendencia

# Mejorar el dise침o visual
fig.update_layout(title_font_size=20,
                  xaxis_title_font_size=14,
                  yaxis_title_font_size=14,
                  template="simple_white")  # Fondo blanco para mayor claridad
fig.show()
```

    <li><strong>Participaci칩n de mercado (%):</strong> [30, 25, 20, 15, 10]</li>
</ul>


```python
# Crear un DataFrame de ejemplo con datos de participaci칩n de mercado de productos
data = {'Marcas': ['Toyota', 'Ford', 'Honda', 'Chevrolet', 'BMW'],
        'Participaci칩n de mercado (%)': [30, 25, 20, 15, 10]} # Participaci칩n de mercado (%): [30, 25, 20, 15, 10]]  # Cantidad de unidades de cada producto


df5 = pd.DataFrame(data)
df5
```

```python
px.pie(df5,
       names="Marcas",
       values="Participaci칩n de mercado (%)")
```

```python
fig =px.pie(df5,
       names="Marcas",
       values="Participaci칩n de mercado (%)",
        title='Participacion de mercado de las Marcas',
        color_discrete_sequence = px.colors.sequential.RdBu)
fig.show()
```

```python
fig =px.pie(df5,
       names="Marcas",
       values="Participaci칩n de mercado (%)",
        title='Participacion de mercado de las Marcas',
        color_discrete_sequence = px.colors.sequential.RdBu)

# Mejorar el dise침o del gr치fico
fig.update_traces(textposition='inside',
                  textinfo='percent+label')  # Mostrar etiquetas y porcentajes en cada secci칩n


# Ajustar la visualizaci칩n del t칤tulo y estilo del gr치fico
fig.update_layout(title_font_size=20,
                  template="plotly_white"   # Fondo blanco para mayor claridad
)
fig.show()
```

 ## **<font color="DarkBlue">Ejercicio 4: Histograma</font>**

<p><strong>Consigna:</strong> Utiliza los siguientes datos sobre los precios de casas en una ciudad para crear un histograma que muestre la distribuci칩n de los precios.</p>
<ul>
    <li><strong>Precios de casas:</strong> [250000, 300000, 350000, 200000, 400000, 450000, 300000, 350000, 270000, 300000, 320000, 380000]</li>
</ul>


```python
data = ({"Precios de casas": [250000, 300000, 350000, 200000, 400000, 450000, 300000, 350000, 270000, 300000, 320000, 380000]})  # Tiempos de entrega en d칤as


df2 = pd.DataFrame(data)
df2
```

```python
# Crear el histograma
fig = px.histogram(df2, x='Precios de casas', nbins=25,
                   title='Precios de las casas',
                   labels={'Precios de casas': 'Precio de casas ($)'},
                   opacity=0.75)
fig.show()
```

```python
# Crear el histograma
fig = px.histogram(df2, x='Precios de casas', nbins=25,
                   title='Precios de las casas',
                   labels={'Precios de casas': 'Precio de casas ($)'},
                   opacity=0.75)

# Mejorar el dise침o del gr치fico
fig.update_layout(
    title_font_size=20,
    xaxis_title_font_size=16,
    yaxis_title_font_size=16,
    template="plotly_white",  # Fondo blanco para mayor claridad
    bargap=0.2                # Separaci칩n entre barras
)
fig.show()
```

```python
import numpy as np
```

```python
# Crear el histograma
fig = px.histogram(df2, x='Precios de casas', nbins=25,
                   title='Precios de las casas',
                   labels={'Precios de casas': 'Precio de casas ($)'},
                   opacity=0.75)
# A침adir l칤nea de densidad para una mejor visualizaci칩n
fig.update_traces(marker_color='cornflowerblue', marker_line_color='darkblue', marker_line_width=1)
fig.add_scatter(x=np.linspace(df2['Precios de casas'].min(), df2['Precios de casas'].max(), 100),
                y=np.exp(-0.5*((np.linspace(df2['Precios de casas'].min(), df2['Precios de casas'].max(), 100) - df2['Precios de casas'].mean()) / df2['Precios de casas'].std())**2) / (df2['Precios de casas'].std() * np.sqrt(2 * np.pi)),
                mode='lines', name='Densidad', line=dict(color='darkorange', width=2))


# Mejorar el dise침o del gr치fico
fig.update_layout(
    title_font_size=20,
    xaxis_title_font_size=16,
    yaxis_title_font_size=16,
    template="plotly_white",  # Fondo blanco para mayor claridad
    bargap=0.2                # Separaci칩n entre barras
)
fig.show()
```

 ## **<font color="DarkBlue">Ejercicio 5: BoxPlot</font>**

<p><strong>Consigna:</strong> Utiliza los siguientes datos de la duraci칩n de llamadas telef칩nicas en minutos para crear un box plot que resuma las duraciones.</p>
<ul>
    <li><strong>Duraci칩n de llamadas (minutos):</strong> [5, 15, 20, 35, 10, 45, 50, 30, 25, 15, 10, 60]</li>
</ul>



```python
data = ({"Duraci칩n de llamada":[5, 15, 20, 35, 10, 45, 50, 30, 25, 15, 10, 60]})  # Tiempos de entrega en d칤as


df3 = pd.DataFrame(data)
df3
```

```python
# Crear el gr치fico box plot
fig = px.box(df3, y='Duraci칩n de llamada',
             title='Duraci칩n de las llamadas telef칩nicas',
             labels={'Duraci칩n de llamada': 'Duraci칩n de las llamadas (minutos)'},
             points='all',
             #color=['Duracion de llamada']) # Removing this line, as it's causing the error
             ) # Removing the color argument altogether solves the problem
# If you want to color the boxplot, you need to provide a list or array with the same length as the data.
# For instance, if you have a column in your dataframe that you want to use for coloring, you could use:
# color='column_name'
# Or if you want to color all boxes the same color, you could use:
# color_discrete_sequence=['cornflowerblue']  # replace with your desired color
fig.show()
```

```python
# Crear el gr치fico box plot
fig = px.box(df3, y='Duraci칩n de llamada',
             title='Duraci칩n de las llamadas telef칩nicas',
             labels={'Duraci칩n de llamada': 'Duraci칩n de las llamadas (minutos)'},
             points='all',
            color_discrete_sequence=['blue'],
             ) # Removing the color argument altogether solves the problem
# Mejorar el dise침o del gr치fico
fig.update_layout(
    title_font_size=18,
    xaxis_title_font_size=16,
    yaxis_title_font_size=16,
    template="plotly_white",  # Fondo blanco para mayor claridad
    showlegend=False
)


fig.show()
```
