 # **<font color="DarkBlue">Agrupaci칩n, groupby</font>**

<p align="justify">Hadley Wickham, autor de muchos paquetes populares para el lenguaje de programaci칩n R, acu침칩 el t칠rmino "dividir-aplicar-combinar" para describir las operaciones de agrupaci칩n.
<br><br>
Este proceso consta de tres etapas: primero, los datos contenidos en un objeto de pandas, como una serie, un DataFrame u otro tipo de objeto, se dividen en grupos basados en una o m치s claves que se especifiquen. Esta divisi칩n se realiza sobre un eje particular del objeto, por ejemplo, un DataFrame puede agruparse por sus filas (<code>axis="index"</code>) o por sus columnas (<code>axis="columns"</code>).
<br><br>
 Una vez que los datos se han dividido, se aplica una funci칩n a cada grupo, generando un nuevo valor para cada uno.
 <br><br>
 Finalmente, los resultados de la aplicaci칩n de estas funciones se combinan para formar un nuevo objeto de resultados. La estructura del objeto resultante depender치 del tipo de operaci칩n que se realice sobre los datos.</p>


游 Ejemplo:

<p align="justify">
游 Supongamos que tenemos un DataFrame con datos de ventas de una tienda. Cada venta est치 asociada a un producto, su categor칤a, y el monto de la venta. Queremos agrupar las ventas por categor칤a de productos y calcular la suma total y el promedio de ventas por cada categor칤a.



```python
import pandas as pd
import plotly.express as px
```

```python
# Crear DataFrame de ventas

ventas = pd.DataFrame({
    'Producto': ['Laptop', 'Smartphone', 'Camiseta', 'Pantalones', 'Televisor', 'Zapatillas'],
    'Categor칤a': ['Electr칩nica', 'Electr칩nica', 'Ropa', 'Ropa', 'Electr칩nica', 'Ropa'],
    'Monto de Venta': [1200, 800, 250, 400, 1500, 600]
})
```

```python
ventas
```

```python
# Agrupar por categor칤a y calcular la venta total por categor칤a

ventas_por_categoria = ventas.groupby('Categor칤a')['Monto de Venta'].sum()
ventas_por_categoria
```

<p align="justify">
El m칠todo <code>groupby()</code> agrupa las filas del DataFrame seg칰n la columna Categor칤a

```python
ventas_por_categoria = ventas.groupby('Categor칤a')['Monto de Venta'].sum()
ventas_por_categoria
```

```python
px.bar(ventas.groupby('Categor칤a')['Monto de Venta'].sum())
```

```python
px.bar(ventas.groupby('Categor칤a')['Monto de Venta'].sum(),
       template="gridon",
       color = ventas.groupby('Categor칤a')['Monto de Venta'].sum().index,
       title="Ventas totales por categor칤a",
       labels={"color":"Categor칤a", "value":"Monto de venta"})
```

<p align="justify">
El par치metro <code>as_index = False</code> agrupa las filas del DataFrame seg칰n la columna Categor칤a, pero la columna Categor칤a ahora no es un indice, sino que es otra columna m치s.

```python
ventas_por_categoria = ventas.groupby('Categor칤a', as_index=False)['Monto de Venta'].sum()
ventas_por_categoria
```

```python
px.bar(ventas_por_categoria,
       x="Categor칤a",
       y="Monto de Venta",
       template="gridon",
       color = "Categor칤a",
       title="Ventas totales por categor칤a",
       )
```

<p align="justify">
Ahora utilizamos <code>agg()</code> para aplicar varias funciones de agregaci칩n, en este caso

- <code>sum()</code> para calcular el total de ventas por categor칤a y
- <code>mean()</code> para calcular el promedio de ventas.


```python
# Agrupar por categor칤a y calcular estad칤sticas

ventas_por_categoria = ventas.groupby('Categor칤a').agg(Total_Ventas = ('Monto de Venta', 'sum'),
                                                       Promedio_Ventas=('Monto de Venta', 'mean')
)
```

```python
ventas_por_categoria
```

```python
# Redondear Promedio_Ventas a dos decimales

ventas_por_categoria['Promedio_Ventas'] = ventas_por_categoria['Promedio_Ventas'].round(2)
```

```python
ventas_por_categoria
```

```python
ventas_por_categoria.reset_index(inplace=True)
ventas_por_categoria
```

```python
fig = px.bar(ventas_por_categoria,
             x = "Categor칤a",
             y = "Total_Ventas",
             template="gridon",
             color="Categor칤a",
             title="Ventas totales por categor칤a con promedio")

fig.add_scatter(x = ventas_por_categoria['Categor칤a'],
                y = ventas_por_categoria['Promedio_Ventas'],
                mode='lines',
                name='Promedio')

fig.show()
```

游 Otro ejemplo:

<p align="justify">
游 Supongamos que tienes un conjunto de datos de ventas de una empresa que opera en distintas regiones y vende diferentes productos. Se analizar치 el promedio de ventas, el total de ventas y la venta m치xima de cada producto en cada regi칩n.




```python
# Datos

data = {
    'Region': ['Norte', 'Norte', 'Norte', 'Sur', 'Sur', 'Sur', 'Este', 'Este', 'Oeste', 'Oeste'],
    'Producto': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'A', 'C'],
    'Ventas': [5000, 6000, 7000, 4000, 3500, 3000, 5500, 6200, 4300, 7800]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# Agrupar por regi칩n y producto

grouped = df.groupby(['Region', 'Producto'])
```

```python
# Operaciones de agregaci칩n

agg_df = grouped.agg(
    Total_Ventas=('Ventas', 'sum'),
    Promedio_Ventas=('Ventas', lambda x: round(x.mean(), 2)), # Funci칩n lambda
    Ventas_Maximas=('Ventas', 'max')
).reset_index()  # Restablecer el 칤ndice para evitar conflictos al agregar columnas
```

```python
agg_df
```

```python
px.bar(agg_df,
       x = "Region",
       y = "Total_Ventas",
       color = "Producto",
       text = "Total_Ventas",
       title="Ventas totales por regi칩n y producto",
       template="gridon",
       labels={"Total_Ventas":"Total ventas en $"})
```

```python
px.sunburst(agg_df,
            path=['Region', 'Producto'],
            values='Total_Ventas',
            color='Producto',
            title="Ventas totales por regi칩n y producto",
            template="gridon")
```

 ## **<font color="DarkBlue">Seleccionando una columna o un subconjunto de columnas</font>**

游 Otro ejemplo:

<p align="justify">
游 Supongamos que tienes un DataFrame con informaci칩n de ventas por producto, regi칩n y categor칤a, y deseas agrupar los datos para obtener el total de ventas por regi칩n y categor칤a, seleccionando solo algunas columnas relevantes para este an치lisis.



```python
# Datos

data = {
    'Producto': ['Producto A', 'Producto B', 'Producto C', 'Producto D', 'Producto E', 'Producto F'],
    'Categoria': ['Electr칩nica', 'Electr칩nica', 'Muebles', 'Muebles', 'Electr칩nica', 'Muebles'],
    'Region': ['Norte', 'Sur', 'Norte', 'Sur', 'Norte', 'Sur'],
    'Ventas': [1000, 1500, 700, 1300, 1200, 900],
    'Costo': [400, 600, 300, 500, 500, 300]
}
```

```python
df = pd.DataFrame(data)
df
```

<p align="justify">
El DataFrame contiene productos, su categor칤a, la regi칩n donde fueron vendidos, las ventas y los costos.

```python
# Agrupar por 'Categoria' y 'Region', y seleccionar solo las columnas 'Ventas' y 'Costo' para hacer la suma

grouped = df.groupby(['Categoria', 'Region'])[['Ventas', 'Costo']].sum()
```

<p align="justify">
Aqu칤 estamos agrupando el DataFrame por dos columnas: Categoria y Region. Luego seleccionamos un subconjunto de columnas, Ventas y Costo, para calcular la suma total de ventas y costos por cada combinaci칩n de categor칤a y regi칩n.

```python
grouped
```

<p align="justify">
Este enfoque es 칰til cuando quieres realizar an치lisis espec칤ficos en un subconjunto de columnas, sin involucrar todas las columnas del DataFrame.



 ## **<font color="DarkBlue">Agrupando con diccionarios y series</font>**

游 Ejemplo:

<p align="justify">
游 Supongamos que tienes un DataFrame que contiene informaci칩n de productos, regiones, ventas y descuentos aplicados, y quieres agrupar las ventas por regi칩n y producto, aplicando un descuento espec칤fico seg칰n el producto usando un diccionario o serie.


```python
# Datos

data = {
    'Producto': ['Producto A', 'Producto B', 'Producto C', 'Producto D', 'Producto E'],
    'Region': ['Norte', 'Sur', 'Norte', 'Sur', 'Norte'],
    'Ventas': [1000, 1500, 700, 1300, 1200],
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# Diccionario de descuentos por producto

descuentos = {
    'Producto A': 0.10,  # 10% de descuento
    'Producto B': 0.15,  # 15% de descuento
    'Producto C': 0.05,  # 5% de descuento
    'Producto D': 0.20,  # 20% de descuento
    'Producto E': 0.10   # 10% de descuento
}
```

<p align="justify">
Hemos creado un diccionario que asigna un descuento a cada producto.

```python
# Crear una serie a partir del diccionario de descuentos

serie_descuentos = pd.Series(descuentos)
serie_descuentos.name = 'Descuentos'
```

<p align="justify">
Convertimos el diccionario en una serie de Pandas para poder aplicar los descuentos de manera eficiente utilizando el m칠todo <code>map().

```python
serie_descuentos
```

```python
# Aplicar los descuentos al DataFrame utilizando groupby

df['Descuento'] = df['Producto'].map(serie_descuentos)
df['Ventas con Descuento'] = df['Ventas'] * (1 - df['Descuento'])
```

<p align="justify">
Utilizamos <code>map()</code> para crear una nueva columna llamada Descuento, que contiene el porcentaje de descuento para cada producto. Luego calculamos las ventas con descuento aplicando la f칩rmula Ventas * (1 - Descuento).

```python
df
```

```python
# Agrupar por 'Region' y 'Producto', y sumar las 'Ventas' y las 'Ventas con Descuento'

grouped = df.groupby(['Region', 'Producto'])[['Ventas', 'Ventas con Descuento']].sum()
```

<p align="justify">
Finalmente, agrupamos los datos por Region y Producto, y sumamos las columnas Ventas y Ventas con Descuento para cada combinaci칩n de regi칩n y producto

```python
grouped
```

 ## **<font color="DarkBlue">Agrupando con funciones</font>**

游 Ejemplo:

<p align="justify">
游 Supongamos que tienes un DataFrame con informaci칩n sobre las ventas de diferentes productos en varias regiones y deseas calcular el total de ventas y el promedio de ventas por producto y regi칩n. Adem치s, aplicaremos una funci칩n personalizada para determinar el rendimiento de cada producto basado en sus ventas.



```python
# Datos

data = {
    'Producto': ['Producto A', 'Producto B', 'Producto A', 'Producto C', 'Producto B', 'Producto C', 'Producto A', 'Producto C'],
    'Region': ['Norte', 'Sur', 'Norte', 'Norte', 'Sur', 'Sur', 'Norte', 'Sur'],
    'Ventas': [2000, 1500, 3000, 4000, 2500, 1500, 3500, 2000],
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# Funci칩n personalizada para evaluar el rendimiento

def evaluar_rendimiento(ventas):
    if ventas.sum() > 8000:
        return 'Alto'
    elif ventas.sum() > 4000:
        return 'Medio'
    else:
        return 'Bajo'
```

<p align="justify">
Definimos la funci칩n evaluar_rendimiento, que eval칰a el rendimiento total de las ventas de un producto y lo clasifica como "Alto", "Medio" o "Bajo" seg칰n las sumas de ventas.

```python
# Agrupar por 'Producto' y 'Region', y calcular el total y el promedio de ventas

grouped = df.groupby(['Producto', 'Region']).agg(
    Total_Ventas=('Ventas', 'sum'),
    Promedio_Ventas=('Ventas', 'mean'),
    Rendimiento=('Ventas', evaluar_rendimiento)  # Aplicar la funci칩n personalizada
).reset_index()
```

```python
grouped.Promedio_Ventas = grouped.Promedio_Ventas.round(2)
```

```python
grouped
```

 ## **<font color="DarkBlue">Agrupando por niveles del 칤ndice</font>**

游 Ejemplo:

<p align="justify">
游 Supongamos que tienes un DataFrame que contiene datos de ventas por producto en diferentes a침os. Primero, crearemos un 칤ndice jer치rquico y luego agruparemos por niveles de 칤ndice.



```python
# Datos

data = {
    'Producto': ['Producto A', 'Producto A', 'Producto B', 'Producto B', 'Producto C', 'Producto C'],
    'A침o': [2021, 2022, 2021, 2022, 2021, 2022],
    'Ventas': [1500, 2000, 3000, 2500, 4000, 3500]
}

```

```python
df = pd.DataFrame(data)
df
```

```python
# Configurar un 칤ndice jer치rquico

df.set_index(['Producto', 'A침o'], inplace=True)
```

<p align="justify">
Usamos <code>set_index()</code> para establecer un 칤ndice jer치rquico con Producto y A침o. Esto nos permite trabajar con m칰ltiples niveles de 칤ndice.

```python
df
```

```python
# Agrupar por el nivel del 칤ndice 'Producto' y sumar las ventas

grouped = df.groupby(level='Producto').sum()
```

Utilizamos <code>groupby(level='Producto')</code> para agrupar el DataFrame por el nivel del 칤ndice correspondiente al Producto. Luego, aplicamos <code>sum()</code> para calcular el total de ventas de cada producto

```python
grouped
```

<p align="justify">
Puedes aplicar diversas funciones de agregaci칩n, como <code>mean(), max(), min()</code>, etc., para obtener diferentes estad칤sticas por grupo.

 # **<font color="DarkBlue">Tablas din치micas y tabulaci칩n cruzada</font>**

<p align="justify">
Una tabla din치mica es una herramienta de resumen de datos que se encuentra con frecuencia en programas de hojas de c치lculo y otro software de an치lisis de datos.
<br><br>
 Podemos agregar una tabla de datos mediante una o m치s claves, organizando los datos en un rect치ngulo con algunas de las claves de grupo a lo largo de las filas y otras a lo largo de las columnas. Las tablas din치micas con pandas son posibles gracias a la funci칩n groupby, combinada con operaciones de remodelaci칩n que utilizan indexaci칩n jer치rquica.
 <br><br>
 DataFrame tambi칠n tiene un m칠todo pivot_table y tambi칠n hay una funci칩n pivot_table de nivel superior.

游 Ejemplo:

<p align="justify">
游 Aqu칤 hay un ejemplo de c칩mo usar groupby y pivot_table en un contexto de an치lisis de ventas. Este ejemplo se centra en un conjunto de datos de ventas de productos y muestra c칩mo agrupar los datos y luego usar una tabla din치mica para resumir la informaci칩n.



```python
# Datos

data = {
    'Fecha': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-03', '2023-01-03'],
    'Producto': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Cantidad': [10, 5, 8, 7, 15, 6],
    'Precio': [100, 200, 100, 200, 100, 200]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# Usar groupby para calcular las ventas totales por producto

df['Ventas Totales'] = df['Cantidad'] * df['Precio']
ventas_por_producto = df.groupby('Producto')['Ventas Totales'].sum().reset_index()
```

```python
ventas_por_producto
```

```python
# Usar pivot_table para obtener una tabla din치mica

tabla_dinamica = df.pivot_table(
    values='Ventas Totales',
    index='Fecha',
    columns='Producto',
    aggfunc='sum',
    fill_value=0
)
```

<p align="justify">
Se crea una tabla din치mica usando pivot_table para resumir las Ventas Totales, con Fecha como 칤ndice y Producto como columnas, usando la funci칩n de agregaci칩n sum.
<br><br>
Se especifica fill_value=0 para rellenar los valores vac칤os con cero.


```python
tabla_dinamica
```
