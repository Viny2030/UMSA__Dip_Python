 # **<font color="DarkBlue">Estad칤sticas descriptivas comunes</font>**

<p align="justify">
Resumen y c치lculo de estad칤sticas descriptivas son fundamentales en el an치lisis de datos, ya que permiten obtener una visi칩n general y cuantitativa de los datos. Estas estad칤sticas ayudan a identificar patrones, tendencias y posibles anomal칤as, facilitando la toma de decisiones informadas.




```python
import pandas as pd
```

```python
# Datos de ventas por producto y regi칩n
data = {
    'Regi칩n': ['Norte', 'Sur', 'Este', 'Oeste'],
    'Producto A': [200, 150, 180, 130],
    'Producto B': [210, 160, 170, 140],
    'Producto C': [180, 120, 160, 150],
    'Producto D': [190, 155, 140, 145]
}
```

```python
# Creando el DataFrame
df = pd.DataFrame(data)
df
```

 ## **<font color="DarkBlue">Res칰men estad칤stico</font>**

```python
# Generar un resumen estad칤stico
df_stats = df.iloc[:, 1:].describe().round(2)
df_stats
```

```python
# Transponer el DataFrame
df_stats.T
```

<p align="justify">
Adem치s de <code>describe()</code>, es posible calcular estad칤sticas espec칤ficas utilizando m칠todos de Pandas:



 ## **<font color="DarkBlue">Media</font>**

```python
# Calcular la media de las ventas por producto
media_ventas = df.iloc[:, 1:].mean()
media_ventas.name = 'Media'
media_ventas
```

 ## **<font color="DarkBlue">Desviaci칩n est치ndar</font>**

```python
# Calcular la desviaci칩n est치ndar de las ventas por producto
std_ventas = df.iloc[:, 1:].std().round(3)
std_ventas.name = 'Desviaci칩n est치ndar'
std_ventas
```

 ## **<font color="DarkBlue">Mediana</font>**

```python
# Calcular la mediana de las ventas por producto
mediana_ventas = df.iloc[:, 1:].median()
mediana_ventas.name = 'Mediana'
mediana_ventas
```

 ## **<font color="DarkBlue">M치ximo</font>**

```python
# Calcular el valor m치ximo de las ventas por producto
max_ventas = df.iloc[:, 1:].max()
max_ventas.name = 'M치ximo'
max_ventas
```

 ## **<font color="DarkBlue">M칤nimo</font>**

```python
# Calcular el valor m칤nimo de las ventas por producto
max_ventas = df.iloc[:, 1:].min()
max_ventas.name = 'M칤nimo'
max_ventas
```

 ## **<font color="DarkBlue">Percentil 25</font>**

```python
# Calcular el percentil 25 de las ventas por producto
percentil_25 = df.iloc[:, 1:].quantile(0.25)
percentil_25.name = 'Percentil 25'
percentil_25
```

<p align="justify">
El uso de estas estad칤sticas descriptivas proporciona a la empresa un entendimiento profundo de c칩mo se est치n comportando las ventas de cada producto en diferentes regiones.
<br><br>
Por ejemplo:
<br><br>

- **Media**: Proporciona una idea del rendimiento promedio.
- **Desviaci칩n est치ndar**: Indica qu칠 tan variable es el rendimiento.
- **Mediana**: Muestra el punto medio de las ventas.
- **M치ximos y m칤nimos**: Ayudan a identificar los mejores y peores resultados.
- **Percentiles**: Ofrecen una visi칩n m치s detallada de la distribuci칩n de las ventas.

 # **<font color="DarkBlue">Correlaci칩n y Covarianza</font>**

<p align="justify">
Conceptos estad칤sticos fundamentales que se utilizan para medir la relaci칩n entre dos variables.


 ## **<font color="DarkBlue">Covarianza</font>**

<p align="justify">
La covarianza mide la direcci칩n de la relaci칩n lineal entre dos variables. Puede ser positiva, negativa o nula:
<br><br>

- **Covarianza positiva**: Si las dos variables tienden a moverse en la misma direcci칩n (cuando una aumenta, la otra tambi칠n aumenta).
- **Covarianza negativa**: Si las dos variables tienden a moverse en direcciones opuestas (cuando una aumenta, la otra disminuye).
- **Covarianza nula**: Si no hay una relaci칩n lineal entre las dos variables.

<p align="justify">
<br>
La covarianza se calcula utilizando la siguiente f칩rmula:
<br><br>
$$
\text{Cov}(X, Y) = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})
$$

Donde:
- $ X $ e $ Y $ son las variables.
- $ \bar{X} $ y $ \bar{Y} $ son las medias de las variables $ X $ e $ Y $.
- $ n $ es el n칰mero de observaciones.

<p align="justify">
<br>
Nota: La covarianza est치 en las unidades del producto de las variables, lo que puede dificultar la interpretaci칩n directa.


<p align="justify">
Supongamos que una empresa quiere analizar la relaci칩n entre sus gastos en marketing y las ventas mensuales para determinar si hay una correlaci칩n positiva entre ambas variables.



```python
# Datos de ventas mensuales y gastos en marketing (en miles de d칩lares)
data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio'],
    'Ventas': [200, 220, 250, 270, 300, 320],
    'Marketing': [50, 55, 60, 65, 70, 75]
}
```

```python
# Creando el DataFrame
df = pd.DataFrame(data)
df
```

<p align="justify">
El m칠todo <code>cov()</code> de Pandas calcula la matriz de covarianza para las columnas seleccionadas. En este caso, se extrae el valor de la covarianza entre ventas y marketing.


```python
# Calcular la covarianza entre las ventas y los gastos de marketing
covarianza = df[['Ventas', 'Marketing']].cov().iloc[0, 1]
covarianza
```

<p align="jusfify">
 El valor positivo de la covarianza (430.0) indica que a medida que los gastos en marketing aumentan, las ventas tambi칠n tienden a aumentar. Sin embargo, el valor en s칤 mismo no tiene una interpretaci칩n f치cil debido a las unidades involucradas.



 ## **<font color="DarkBlue">Correlaci칩n</font>**

<p align="justify">
La correlaci칩n mide tanto la direcci칩n como la fuerza de la relaci칩n lineal entre dos variables. A diferencia de la covarianza, la correlaci칩n es adimensional y est치 acotada entre -1 y 1:
<br><br>

- **Correlaci칩n positiva (+1)**: Indica una relaci칩n perfectamente positiva.
- **Correlaci칩n negativa (-1)**: Indica una relaci칩n perfectamente negativa.
- **Correlaci칩n cero (0)**: Indica que no hay una relaci칩n lineal entre las variables.

<p align="justify">
<br><br>
La correlaci칩n se calcula utilizando la siguiente f칩rmula, que normaliza la covarianza:

$$
\text{Corr}(X, Y) = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
$$

Donde:
- $ \sigma_X $ y $ \sigma_Y $ son las desviaciones est치ndar de $ X $ e $ Y $.


<p align="justify">
El m칠todo <code>corr()</code> de Pandas calcula la matriz de correlaci칩n. Similar al caso de la covarianza, se extrae el valor de la correlaci칩n entre las dos variables.


```python
# Calcular la correlaci칩n entre las ventas y los gastos de marketing
correlacion = df[['Ventas', 'Marketing']].corr().iloc[0, 1]
correlacion
```

<p align="justify">
 La correlaci칩n de 0.998 indica una relaci칩n lineal casi perfecta y positiva entre los gastos en marketing y las ventas. Esto significa que el aumento en los gastos de marketing est치 directamente asociado con un aumento proporcional en las ventas.
