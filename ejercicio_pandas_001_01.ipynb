 ## **<font color="DarkBlue">Ejercicio 1: Selección condicional y creación de nuevas columnas</font>**

En este ejercicio, realizaremos una selección condicional para identificar productos con un precio superior a un cierto umbral y crearemos una nueva columna para clasificar los productos.



 ### **<font color="DarkBlue">Consigna</font>**

1. Crea un DataFrame con información sobre productos, incluyendo las columnas 'Producto', 'Precio' y 'Categoría'.
2. Selecciona los productos cuyo precio sea superior a 100.
3. Crea una nueva columna llamada 'Rango_Precio' y clasifica los productos como 'Alto' si el precio es mayor a 150, 'Medio' si está entre 100 y 150, y 'Bajo' si es menor a 100.

```
# Desarrollar la consigna
```

 ### **<font color="DarkBlue">Solucion propuesta</font>**

```
import pandas as pd

# Crear el DataFrame
datos_productos = {'Producto': ['Camiseta', 'Pantalón', 'Zapatilla', 'Gorra', 'Remera'],
                  'Precio': [90, 120, 180, 50, 110],
                  'Categoría': ['Indumentaria', 'Indumentaria', 'Calzado', 'Accesorio', 'Indumentaria']}

df_productos = pd.DataFrame(datos_productos)
df_productos
```

```
# Seleccionar productos con precio > 100
productos_caros = df_productos[df_productos['Precio'] > 100]
productos_caros
```

```
# Crear columna Rango_Precio
def clasificar_precio(precio):
    if precio > 150:
        return 'Alto'
    elif precio >= 100:
        return 'Medio'
    else:
        return 'Bajo'
```

```
df_productos['Rango_Precio'] = df_productos['Precio'].apply(clasificar_precio)
df_productos
```

 ## **<font color="DarkBlue">Ejercicio 2: Filtrado complejo con múltiples condiciones y funciones booleanas</font>**

En este ejercicio, combinaremos múltiples condiciones y utilizaremos funciones booleanas para filtrar los datos de manera más precisa.


 ### **<font color="DarkBlue">Consigna</font>**


1. Crea un DataFrame con información sobre clientes, incluyendo las columnas 'Nombre', 'Edad', 'Ciudad' y 'Cliente_VIP'.
2. Filtra los clientes que sean mayores de 30 años, vivan en 'Buenos Aires' y sean clientes VIP.
3. Utiliza la función `query()` para realizar el mismo filtrado.


```
# Desarrollar consigna
```

 ### **<font color="DarkBlue">Solución propuesta</font>**

```
import pandas as pd

# Crear el DataFrame
datos_clientes = {'Nombre': ['Juan', 'Ana', 'Carlos', 'María'],
                 'Edad': [35, 28, 42, 30],
                 'Ciudad': ['Buenos Aires', 'Córdoba', 'Rosario', 'Buenos Aires'],
                 'Cliente_VIP': [True, False, True, True]}

df_clientes = pd.DataFrame(datos_clientes)
df_clientes
```

```
# Filtrar con múltiples condiciones
clientes_filtrados = df_clientes[(df_clientes['Edad'] > 30) & (df_clientes['Ciudad'] == 'Buenos Aires') & (df_clientes['Cliente_VIP'] == True)]
clientes_filtrados
```

```
# Filtrar con query()
clientes_filtrados_query = df_clientes.query('Edad > 30 and Ciudad == "Buenos Aires" and Cliente_VIP == True')
clientes_filtrados_query
```

 ## **<font color="DarkBlue">Ejercicio 3: Selección por posición y etiquetas</font>**

En este ejercicio, exploraremos diferentes formas de seleccionar filas y columnas utilizando índices numéricos y etiquetas.



 ### **<font color="DarkBlue">Consigna</font>**

1. Crea un DataFrame con datos aleatorios.
2. Selecciona las primeras 3 filas.
3. Selecciona las columnas 'Columna1' y 'Columna3'.
4. Selecciona las filas con índice par.



```
# Desarrollar la consigna
```

 ### **<font color="DarkBlue">Solución propuesta</font>**

```
import pandas as pd
import numpy as np

# Crear DataFrame aleatorio
df = pd.DataFrame(np.random.randn(5, 4), columns=['Columna1', 'Columna2', 'Columna3', 'Columna4'])
df
```

```
# Seleccionar primeras 3 filas
primeras_filas = df.iloc[:3]
primeras_filas
```

```
# Seleccionar columnas específicas
columnas_seleccionadas = df[['Columna1', 'Columna3']]
columnas_seleccionadas
```

```
# Seleccionar filas con índice par
filas_pares = df.iloc[::2]
filas_pares
```
