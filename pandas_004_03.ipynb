 ## **<font color="DarkBlue">Codificaci贸n de categor铆as nominales (sin asumir orden)</font>**

<p align="justify">
 La codificaci贸n one hot es utilizado cuando las categor铆as de una variable no asumen un orden. Para una variable determinada, crear谩 tantas columnas nuevas como clases o valores contenga esa columna. Para una muestra dada, el valor de la columna correspondiente a la categor铆a se establecer谩 en $1$ mientras que todas las columnas de las dem谩s categor铆as se establecer谩n en $0$.

<p align="justify">
 El m茅todo <code>get_dummies</code> en Pandas se utiliza para convertir variables categ贸ricas en variables dummy (ficticia o binaria), que son columnas binarias que indican la presencia o ausencia de una categor铆a particular en un conjunto de datos.

```python
# Codificaci贸n de variables categ贸ricas con pandas
data_categorical_encoded = pd.get_dummies(data_categorical, dtype=int)
data_categorical_encoded.head()
```

<p align="justify">
 El n煤mero de caracter铆sticas despu茅s de la codificaci贸n es $10$ veces mayor que los datos originales porque algunas variables, como la ocupaci贸n y el pa铆s de origen, tienen muchos valores 煤nicos.

 ## **<font color="DarkBlue">Normalizaci贸n de variables num茅ricas</font>**

<p align="justify">
La normalizaci贸n de variables num茅ricas es fundamental en el preprocesamiento de datos, especialmente para algoritmos de aprendizaje autom谩tico. Se puede realizar con Scikit-Learn, Pandas y Numpy. Los datos se ajustan entre un rango espec铆fico (por ejemplo entre 0 y 1) utilizando la siguiente f贸rmula:
$$
X_{\text{normalizado}} = \frac{X - X_{\text{min}}}{X_{\text{max}} - X_{\text{min}}}
$$

```python
# Normalizaci贸n de data_numerical
# Normalizar entre 0 y 1
data_numerical_normalizado = (data_numerical - data_numerical.min()) / (data_numerical.max() - data_numerical.min())
data_numerical_normalizado.head()
```

 ## **<font color="DarkBlue">Concatenaci贸n de DataFrames</font>**

<p align="justify">
Para concatenar DataFrames de forma horizontal (a帽adiendo columnas), se puede utilizar el m茅todo <code>concat</code> de Pandas, con el par谩metro <code>axis=1</code>.

https://pandas.pydata.org/docs/reference/api/pandas.concat.html


```python
# Concatenamos los DataFrames
data_encoded = pd.concat([data_categorical_encoded, data_numerical_normalizado], axis=1)
data_encoded.head()
```
