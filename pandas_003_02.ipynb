 # **<font color="DarkBlue">Transformaci칩n de datos</font>**

<p align="justify">
La transformaci칩n de datos es el proceso de modificar, organizar o estructurar los datos para que sean m치s 칰tiles en un an치lisis o modelado.
<br><br>
Esto puede implicar convertir datos de un formato a otro, aplicar funciones matem치ticas, normalizar, escalar, agregar o dividir columnas, entre otras operaciones.
<br><br>
El objetivo es adaptar los datos para que se ajusten a los requisitos del an치lisis, mejorando la precisi칩n y eficiencia del procesamiento.
<br><br>
En los negocios, es crucial para preparar la informaci칩n antes de aplicar modelos predictivos, realizar an치lisis estad칤sticos o generar reportes.

 ## **<font color="DarkBlue">Eliminando duplicados</font>**

<p align="justify">
Este tipo de an치lisis es fundamental en negocios donde la calidad de los datos de los clientes es crucial.
<br><br>
Al eliminar registros duplicados, las empresas pueden mejorar la precisi칩n de sus campa침as de marketing, an치lisis de comportamiento del cliente y personalizaci칩n de servicios, lo que puede llevar a una mejor experiencia del cliente y a mayores ventas.








```python
import pandas as pd
```

游 Ejemplo:

```python
# Crear un DataFrame simulado con datos de clientes, incluyendo duplicados
data = {
    'Cliente_ID': [1, 2, 2, 3, 4, 5, 5, 6],
    'Nombre': ['Ana', 'Luis', 'Luis', 'Pedro', 'Mar칤a', 'Juan', 'Juan', 'Sof칤a'],
    'Email': ['ana@example.com', 'luis@example.com', 'luis@example.com',
              'pedro@example.com', 'maria@example.com', 'juan@example.com',
              'juan@example.com', 'sofia@example.com'],
    'Telefono': ['111-222-333', '222-333-444', '222-333-444',
                 '333-444-555', '444-555-666', '555-666-777',
                 '555-666-777', '666-777-888']
}
```

```python
# Crear el DataFrame
df_clientes = pd.DataFrame(data)
df_clientes
```

 ### **<font color="DarkBlue">Estrategias de manejo de datos duplicados </font>**

```python
# 1. Identificar duplicados bas치ndose en todas las columnas

duplicados_todos = df_clientes[df_clientes.duplicated()]
print("\nDuplicados (todas las columnas):")
duplicados_todos
```

```python
# 2. Eliminar duplicados bas치ndose en todas las columnas

df_sin_duplicados = df_clientes.drop_duplicates()
print("\nDatos sin duplicados (todas las columnas):")
df_sin_duplicados
```

```python
# 3. Identificar duplicados bas치ndose en un subconjunto de columnas (Email y Telefono)

duplicados_subset = df_clientes[df_clientes.duplicated(subset=['Email', 'Telefono'])]
print("\nDuplicados (Email y Telefono):")
duplicados_subset
```

```python
# 4. Eliminar duplicados bas치ndose en un subconjunto de columnas (Email y Telefono)

df_sin_duplicados_subset = df_clientes.drop_duplicates(subset=['Email', 'Telefono'])
print("\nDatos sin duplicados (Email y Telefono):")
df_sin_duplicados_subset
```

```python
# 5. Mantener la 칰ltima aparici칩n en caso de duplicados

df_mantener_ultimo = df_clientes.drop_duplicates(subset=['Email', 'Telefono'], keep='last')
print("\nDatos sin duplicados (mantener 칰ltima aparici칩n - Email y Telefono):")
df_mantener_ultimo
```

 ## **<font color="DarkBlue">Transformaci칩n mediante funci칩n o mapeo </font>**

<p align="justify">
Para muchos conjuntos de datos, es posible que desee realizar alguna transformaci칩n en funci칩n de los valores de una matriz, una serie o una columna en un DataFrame

```python
# Crear un DataFrame simulado con datos de ventas
data = {
    'Producto': ['Laptop', 'Smartphone', 'Tablet', 'Smartwatch', 'Desktop'],
    'Precio': [1000, 500, 300, 200, 1500],  # precios en d칩lares
    'Categoria': ['Electronics', 'Electronics', 'Electronics', 'Wearables', 'Electronics']
}
```

```python
# Crear el DataFrame
df_ventas = pd.DataFrame(data)
df_ventas
```

 ### **<font color="DarkBlue">Mediante funci칩n</font>**

```python
# Definir una funci칩n para aplicar un impuesto al precio

def aplicar_iva(precio, tasa_impuesto=0.21):
    """Aplica una tasa de impuesto a un precio dado."""
    return precio * (1 + tasa_impuesto)
```

```python
# Usar la funci칩n para transformar la columna 'Precio'

df_ventas['Precio_Con_Impuesto'] = df_ventas['Precio'].apply(aplicar_iva)
print("\nDatos con precios transformados (con impuesto):")
df_ventas
```

 ### **<font color="DarkBlue">Mediante mapeo</font>**

```python
# Definir un mapeo para simplificar las categor칤as

mapeo_categorias = {
    'Electronics': 'Electr칩nica',
    'Wearables': 'Vestibles'
}
```

```python
# Usar el mapeo para transformar la columna 'Categoria'

df_ventas['Categoria_Transformada'] = df_ventas['Categoria'].map(mapeo_categorias)
print("\nDatos con categor칤as transformadas:")
df_ventas
```

```python
# Limpiar el DataFrame eliminando columnas innecesarias

df_ventas_final = df_ventas.drop(columns=['Precio', 'Categoria'])
print("\nDatos finales de ventas:")
df_ventas_final
```

 ## **<font color="DarkBlue">Discretizaci칩n de datos y agrupamiento</font>**

<p align="justify">
Los datos continuos suelen discretizarse o separarse en 랂rupos para su an치lisis.


```python
# Crear un DataFrame simulado con datos de ventas de productos

data = {
    'Producto': ['Laptop', 'Smartphone', 'Tablet', 'Smartwatch', 'Desktop', 'C치mara', 'Auriculares', 'Impresora'],
    'Ventas': [1500, 1200, 800, 600, 2500, 300, 700, 1000]  # Ventas en unidades
}
```

```python
# Crear el DataFrame
df_ventas = pd.DataFrame(data)
df_ventas
```

 ### **<font color="DarkBlue">Estrategias para la discretizaci칩n de datos y agrupamiento </font>**

```python
# Definir los l칤mites para las categor칤as de ventas

bins = [0, 500, 1000, 2000, 3000]
labels = ['Bajo', 'Promedio', 'Alto', 'Muy Alto']
```

```python
# Usar pandas.cut() para clasificar las ventas

df_ventas['Rendimiento_Ventas'] = pd.cut(df_ventas['Ventas'], bins=bins, labels=labels, right=False)
```

```python
# Mostrar el DataFrame con la clasificaci칩n de rendimiento de ventas

print("\nDatos con clasificaci칩n de rendimiento de ventas:")
df_ventas
```

```python
# Contar la cantidad de productos en cada categor칤a

conteo_rendimiento = df_ventas['Rendimiento_Ventas'].value_counts()
print("\nConteo de productos por categor칤a de rendimiento de ventas:")
conteo_rendimiento.name = 'Cantidad'
conteo_rendimiento
```

```python
# Mostrar productos que necesitan atenci칩n (Bajo rendimiento)

productos_bajo_rendimiento = df_ventas[df_ventas['Rendimiento_Ventas'] == 'Bajo']
print("\nProductos que necesitan atenci칩n (Bajo rendimiento):")
productos_bajo_rendimiento
```

 ## **<font color="DarkBlue">Detecci칩n de valores at칤picos</font>**

<p align="justify">
Imaginemos que tenemos un conjunto de datos sobre gastos mensuales de diferentes empresas en un sector espec칤fico.
<br><br>
Queremos identificar aquellos gastos que son an칩malos (valores at칤picos) que pueden requerir una revisi칩n m치s cercana, como posibles errores de entrada de datos o gastos inusuales.



```python
import numpy as np
import plotly.express as px
```

```python
# Crear un DataFrame simulado con gastos mensuales de empresas
data = {
    'Empresa': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L'],
    'Gastos Mensuales': [12000, 15000, 11000, 13000, 25000, 14000, 16000, 15500, 13500, 12500, 9500, 21500]  # Gastos en d칩lares
}
```

```python
df_gastos = pd.DataFrame(data)
df_gastos
```

```python
# Calcular el rango intercuart칤lico (IQR)

Q1 = df_gastos['Gastos Mensuales'].quantile(0.25)
Q3 = df_gastos['Gastos Mensuales'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Definir los l칤mites para identificar valores at칤picos

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
```

```python
# Identificar los valores at칤picos

df_gastos['Anomalo'] = (df_gastos['Gastos Mensuales'] < limite_inferior) | (df_gastos['Gastos Mensuales'] > limite_superior)
```

```python
# Mostrar el DataFrame con la detecci칩n de valores at칤picos

print("\nDatos con detecci칩n de valores at칤picos:")
df_gastos
```

```python
# Visualizar los gastos y los valores at칤picos con Plotly

px.box(df_gastos,
       y='Gastos Mensuales',
       title='Boxplot de Gastos Mensuales',
       points="all")
```

```python
# Filtrar los valores at칤picos

valores_atipicos = df_gastos[df_gastos['Anomalo']]
print("\nValores at칤picos detectados:")
valores_atipicos
```

 ## **<font color="DarkBlue">Variables ficticias</font>**

<p align="justify">
Otro tipo de transformaci칩n para aplicaciones de modelado estad칤stico o aprendizaje autom치tico es convertir una variable categ칩rica en una matriz ficticia.
<br><br>
El uso de get_dummies() es fundamental en el an치lisis de datos econ칩micos, especialmente en modelos de regresi칩n y an치lisis multivariante, donde es crucial convertir variables categ칩ricas en un formato que los algoritmos de aprendizaje autom치tico puedan procesar.
<br><br>
Este enfoque permite a las empresas comprender mejor el impacto de las diferentes categor칤as de productos en sus ventas, facilitando decisiones estrat칠gicas sobre marketing, inventario y desarrollo de productos.








```python
# Crear un DataFrame simulado con datos de ventas

data = {
    'Producto': ['Laptop', 'Camisa', 'Manzana', 'Televisor', 'Zapatos', 'Pl치tano', 'Tablet', 'Pantalones', 'Cereal', 'C치mara'],
    'Categoria': ['Electr칩nica', 'Ropa', 'Alimentos', 'Electr칩nica', 'Ropa', 'Alimentos', 'Electr칩nica', 'Ropa', 'Alimentos', 'Electr칩nica'],
    'Ventas': [1200, 300, 150, 800, 500, 100, 450, 250, 350, 600]
}

df_ventas = pd.DataFrame(data)
df_ventas
```

```python
# Usar get_dummies() para convertir la columna 'Categoria' en variables dummy

df_dummies = pd.get_dummies(df_ventas, columns=['Categoria'])
print("\nDatos de ventas con variables dummy:")
df_dummies
```

```python
# Usar get_dummies() para convertir la columna 'Categoria' en variables dummy

df_dummies = pd.get_dummies(df_ventas, columns=['Categoria'], drop_first=True)
df_dummies
```

<p align="justify">
Utilizamos pd.get_dummies() para convertir la columna Categoria en variables dummy.
<br><br>
La opci칩n drop_first=True se utiliza para evitar la multicolinealidad al eliminar la primera columna dummy, ya que es redundante.
<br><br>
Esto crea nuevas columnas para cada categor칤a, donde cada columna indica la presencia (1) o ausencia (0) de esa categor칤a para cada producto
