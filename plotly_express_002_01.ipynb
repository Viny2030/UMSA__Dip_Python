Un gr치fico de <i>barras de errores</i> muestra valores de datos con su rango de error o variabilidad en torno al valor central. Se utilizan para representar la incertidumbre o variabilidad en una medida, permitiendo la evaluaci칩n de la confiabilidad de los resultados. <br><br>

<br>
Este gr치fico es 칰til en an치lisis comparativos, donde queremos saber si la variaci칩n observada en los datos de diferentes grupos es significativa.

<br><br>

Este gr치fico es ideal para identificar la dispersi칩n, la simetr칤a de la distribuci칩n y los posibles valores at칤picos en los datos.

 # **<font color="DarkBlue">Generando un gr치fico Barra de Errores</font>**

<p align="justify">
Un <b>gr치fico de barras de errores</b> es una herramienta de visualizaci칩n que muestra el valor promedio de una variable junto con una barra que indica su rango de variabilidad o incertidumbre. Cada barra de error representa un intervalo alrededor del valor promedio, el cual puede estar basado en el error est치ndar, la desviaci칩n est치ndar o el intervalo de confianza de los datos. Este gr치fico permite visualizar si las diferencias observadas entre categor칤as son significativas y si la variabilidad en los datos es considerable.
<br><br>
La barra de error puede calcularse de varias formas seg칰n el contexto:
<ul>
    <li>El <i>error est치ndar de la media</i> se usa para representar la variabilidad de una muestra en relaci칩n con la media, calculado como $\frac{s}{\sqrt{n}}$, donde $s$ es la desviaci칩n est치ndar y $n$ el tama침o de la muestra.</li>
    <li>La <i>desviaci칩n est치ndar</i> se utiliza para representar la dispersi칩n de los datos alrededor de la media.</li>
    <li>El <i>intervalo de confianza</i> se utiliza para representar un rango en el cual se espera que el valor verdadero se encuentre con una determinada probabilidad, generalmente 95%.</li>
</ul>
<p align="justify">
<br>
En el an치lisis de negocios, los gr치ficos de barras de errores son 칰tiles para comparar variables en diferentes grupos, como ventas promedio en distintas regiones, y evaluar la confianza en los resultados, ayudando a identificar variaciones significativas y tomar decisiones basadas en la confiabilidad de los datos.</p>


<p align="justify"> 游 Ejemplo:  
<br><br>
Gr치fico de barras de errores para visualizar el promedio de ventas en diferentes regiones y su variabilidad.

```python
# Importar las librer칤as necesarias
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
# Crear un DataFrame de ejemplo con datos de ventas y desviaci칩n est치ndar en diferentes regiones
data = {'Region': ['Norte', 'Sur', 'Este', 'Oeste'],
        'Ventas_Promedio': [20000, 25000, 22000, 27000],  # Ventas promedio en d칩lares
        'Desviacion': [3000, 4000, 3500, 4500]            # Desviaci칩n est치ndar de ventas
}

df = pd.DataFrame(data)
df
```

```python
# Crear el gr치fico de barras de errores
fig = px.bar(df,
             x='Region',
             y='Ventas_Promedio',
             title='Promedio de Ventas por Regi칩n con Barras de Error',
             labels={'Ventas_Promedio': 'Ventas Promedio ($)', 'Region': 'Regi칩n'},
             error_y='Desviacion',  # A침adir la barra de error usando la desviaci칩n est치ndar
             color='Region')


# Mejorar el dise침o del gr치fico
fig.update_layout(
    template="plotly_white",  # Fondo blanco para mayor claridad
    showlegend=False
)


# Mostrar el gr치fico
fig.show()
```

En este gr치fico de barras de errores:

- Las barras representan el promedio de ventas para cada regi칩n.
- Las barras de error reflejan la desviaci칩n est치ndar de ventas en cada regi칩n, proporcionando una idea de la variabilidad de las ventas.

<p align="justify">
<br>
Este gr치fico ayuda a identificar si la variabilidad en las ventas es alta o baja en cada regi칩n, permitiendo a los analistas de negocios evaluar si las diferencias en el promedio de ventas son estad칤sticamente significativas y si es necesario tomar acciones para mejorar la estabilidad de las ventas en ciertas regiones.

```python
# Crear el gr치fico de barra de errores
fig = px.line(df,
              x='Region',
              y='Ventas_Promedio',
              title='Promedio de Ventas por Regi칩n con Barras de Error',
              labels={'Ventas_Promedio': 'Ventas Promedio ($)', 'Region': 'Regi칩n'},
              error_y='Desviacion',  # A침adir la barra de error usando la desviaci칩n est치ndar
              markers=True
              )


# Mejorar el dise침o del gr치fico
fig.update_layout(
    template="plotly_white",  # Fondo blanco para mayor claridad
    showlegend=False
)


# Mostrar el gr치fico
fig.show()
```

 # **<font color="DarkBlue">Generando un gr치fico Box Plots</font>**

<p align="justify">
Un <b>box plot</b>, o <i>gr치fico de caja y bigotes</i>, es un gr치fico estad칤stico que representa la distribuci칩n de un conjunto de datos num칠ricos mediante un resumen de cinco n칰meros clave: el m칤nimo, el primer cuartil (Q1), la mediana (Q2), el tercer cuartil (Q3) y el m치ximo. Estos elementos se visualizan en una caja (que representa el rango intercuart칤lico o IQR) y dos l칤neas llamadas <i>bigotes</i> que indican la variaci칩n de los datos fuera de los cuartiles. El IQR se calcula como:
<br><br>
$$
IQR = Q3 - Q1
$$
<br>
La caja abarca desde Q1 hasta Q3, donde se encuentra el 50% de los datos. La mediana, ubicada dentro de la caja, divide los datos en dos partes iguales. Los bigotes indican la extensi칩n de los datos, aunque excluyen los valores at칤picos, que suelen mostrarse como puntos individuales fuera de los bigotes.
<br><br>
Los box plots son 칰tiles para observar la distribuci칩n y variabilidad de los datos, identificar valores at칤picos, y comparar la dispersi칩n de datos en diferentes grupos. En un contexto de negocios, los box plots pueden ser utilizados para analizar el rango de variabilidad en ventas, tiempos de entrega, ingresos, entre otros, y observar si hay diferencias significativas entre grupos o categor칤as.</p>


<p align="justify"> 游 Ejemplo:  
<br><br>
Gr치fico de box plot para analizar la distribuci칩n de ventas mensuales en diferentes regiones de una empresa.

```python
# Crear un DataFrame de ejemplo con datos de ventas mensuales en diferentes regiones
np.random.seed(42)  # Para reproducibilidad de datos
data = {'Region': np.repeat(['Norte', 'Sur', 'Este', 'Oeste'], 50),
        'Ventas': np.concatenate([np.random.normal(20000, 3000, 50),  # Ventas en Norte
                                  np.random.normal(25000, 5000, 50),  # Ventas en Sur
                                  np.random.normal(22000, 4000, 50),  # Ventas en Este
                                  np.random.normal(27000, 4500, 50)   # Ventas en Oeste
    ])
}

df = pd.DataFrame(data)
df
```

```python
# Crear el gr치fico box plot
fig = px.box(df,
             x='Region',
             y='Ventas',
             title='Distribuci칩n de Ventas Mensuales por Regi칩n',
             labels={'Ventas': 'Ventas Mensuales ($)', 'Region': 'Regi칩n'},
             color='Region')


# Mejorar el dise침o del gr치fico
fig.update_layout(
    title_font_size=18,
    xaxis_title_font_size=16,
    yaxis_title_font_size=16,
    template="plotly_white",  # Fondo blanco para mayor claridad
    showlegend=False
)

# Mostrar el gr치fico
fig.show()
```

En este box plot:
<br><br>

- Las cajas representan el rango intercuart칤lico (IQR), que abarca el 50% central de las ventas en cada regi칩n.
- La l칤nea en el centro de cada caja es la mediana de las ventas mensuales en cada regi칩n.
- Los bigotes representan la extensi칩n de los datos hasta aproximadamente 1.5 veces el IQR desde Q1 y Q3, y los puntos fuera de los bigotes son valores at칤picos.

<p align="justify">
<br>
Este gr치fico permite observar la variabilidad en las ventas mensuales de cada regi칩n y comparar la dispersi칩n entre ellas. Por ejemplo, si una regi칩n tiene valores at칤picos elevados, podr칤a investigarse para entender si hay factores espec칤ficos que elevan sus ventas en ciertos meses.

 # **<font color="DarkBlue">Generando un Histograma</font>**

<p align="justify">
Un <b>histograma</b> es una representaci칩n gr치fica de la distribuci칩n de un conjunto de datos continuos. Los datos se dividen en intervalos o <i>bins</i>, y cada bin muestra la frecuencia de los valores que caen dentro de ese rango. La altura de cada barra en el histograma indica el n칰mero de observaciones dentro de cada intervalo, permitiendo visualizar la densidad de los datos a lo largo de su rango.
<br><br>
El histograma es particularmente 칰til para observar la forma de la distribuci칩n de los datos (sim칠trica, sesgada, multimodal, etc.), la concentraci칩n de valores y la variabilidad. Esto permite a los analistas identificar patrones como sesgos, acumulaci칩n en ciertos rangos, y detectar valores extremos. Adem치s, los histogramas facilitan el an치lisis de la dispersi칩n, ayudando a visualizar la consistencia de los datos y posibles tendencias en su frecuencia.
<br><br>
En un contexto de negocios, los histogramas se utilizan para observar la distribuci칩n de variables como ingresos, tiempos de entrega, precios de productos, entre otros. Esto puede ayudar a las empresas a identificar rangos de valores con m치s ocurrencias y ajustar sus estrategias en funci칩n de la distribuci칩n observada.</p>


<p align="justify"> 游 Ejemplo:  
<br><br>
Gr치fico histograma para visualizar la distribuci칩n de tiempos de entrega de pedidos en una empresa.

```python
# Crear un DataFrame de ejemplo con datos de tiempos de entrega (en d칤as) de diferentes pedidos
np.random.seed(42)  # Para reproducibilidad de datos
data = {'Tiempo_Entrega': np.random.normal(5, 1.5, 1000)  # Tiempos de entrega en d칤as
}

df = pd.DataFrame(data)
df
```

```python
# Crear el histograma
fig = px.histogram(df, x='Tiempo_Entrega', nbins=20,
                   title='Distribuci칩n de Tiempos de Entrega de Pedidos',
                   labels={'Tiempo_Entrega': 'Tiempo de Entrega (d칤as)'},
                   opacity=0.75)

# Mejorar el dise침o del gr치fico
fig.update_layout(
    title_font_size=20,
    xaxis_title_font_size=16,
    yaxis_title_font_size=16,
    template="plotly_white",  # Fondo blanco para mayor claridad
    bargap=0.2                # Separaci칩n entre barras
)

# Mostrar el gr치fico
fig.show()
```

En este histograma:
<br><br>

- Cada barra representa el n칰mero de pedidos que se entregaron en un rango de d칤as espec칤fico.
- La altura de cada barra muestra la frecuencia de tiempos de entrega dentro de ese rango.

<p align="justify">
<br>
Este histograma permite a la empresa ver cu치l es el tiempo de entrega m치s com칰n y c칩mo se distribuyen los tiempos en general. Si hay una acumulaci칩n de tiempos altos, podr칤a ser necesario investigar retrasos o problemas en el proceso de entrega. Por otro lado, si el tiempo de entrega promedio es bajo y consistente, la empresa podr칤a usar esta m칠trica como un indicador positivo en sus informes de servicio.








```python
# Crear el histograma con ajustes adicionales
fig = px.histogram(df,
                   x='Tiempo_Entrega',
                   nbins=20,
                   title='Distribuci칩n de Tiempos de Entrega de Pedidos',
                   labels={'Tiempo_Entrega': 'Tiempo de Entrega (d칤as)'},
                   opacity=0.7,  # Ajuste de opacidad para claridad en la visualizaci칩n
                   marginal='rug',  # Agrega un "rug plot" para visualizar datos individuales
                   histnorm='probability density'  # Normalizar el histograma para ver la densidad
                  )

# A침adir l칤nea de densidad para una mejor visualizaci칩n
fig.update_traces(marker_color='cornflowerblue', marker_line_color='darkblue', marker_line_width=1)
fig.add_scatter(x=np.linspace(df['Tiempo_Entrega'].min(), df['Tiempo_Entrega'].max(), 100),
                y=np.exp(-0.5*((np.linspace(df['Tiempo_Entrega'].min(), df['Tiempo_Entrega'].max(), 100) - df['Tiempo_Entrega'].mean()) / df['Tiempo_Entrega'].std())**2) / (df['Tiempo_Entrega'].std() * np.sqrt(2 * np.pi)),
                mode='lines', name='Densidad', line=dict(color='darkorange', width=2))

# Mejorar el dise침o del gr치fico
fig.update_layout(
    title_font_size=18,
    xaxis_title_font_size=16,
    yaxis_title_font_size=16,
    template="plotly_white",  # Fondo blanco para mayor claridad
    yaxis_title="Densidad de Frecuencia",
    xaxis=dict(tickangle=-45),
    bargap=0.2
)

# Mostrar el gr치fico
fig.show()
```

- **Normalizaci칩n de Frecuencia**: Al normalizar el histograma como "densidad de probabilidad" (histnorm='probability density'), podemos observar la forma de la distribuci칩n sin que dependa del tama침o de los datos, lo cual permite comparar f치cilmente distribuciones similares.
- **Rug Plot**: Agregar una l칤nea marginal o "rug plot" (marginal='rug') proporciona un desglose detallado de la posici칩n de los datos individuales a lo largo del eje X.
- **L칤nea de Densidad**: Se a침adi칩 una l칤nea de densidad normal te칩rica en color naranja sobre el histograma, lo cual facilita la interpretaci칩n de la distribuci칩n y permite ver si los datos siguen aproximadamente una distribuci칩n normal.
- **Ajustes de Color y Estilo**: Colores contrastantes en la barra y borde de cada intervalo del histograma (marker_color y marker_line_color), mejorando la claridad y atractivo visual.

<br>
<p align="justify">
Este gr치fico permite observar r치pidamente la concentraci칩n de tiempos de entrega, si hay sesgo en los datos y cualquier posible asimetr칤a o variabilidad que pueda estar presente.

 # **<font color="DarkBlue">Generando un Parallel Categories Diagram</font>**

<p align="justify">
Un <b>Parallel Categories Diagram</b>, o diagrama de categor칤as paralelas, es una herramienta de visualizaci칩n que permite analizar relaciones complejas entre varias variables categ칩ricas al mostrarlas en una serie de ejes paralelos. Cada eje representa una variable, y las l칤neas que conectan los ejes muestran combinaciones de valores de las distintas categor칤as, permitiendo explorar c칩mo los valores de una categor칤a est치n relacionados con los valores de otras categor칤as.
<br><br>
Este tipo de diagrama es especialmente 칰til para identificar patrones y flujos en datos con varias dimensiones categ칩ricas, como puede ser la relaci칩n entre caracter칤sticas demogr치ficas de clientes, tipos de productos y comportamientos de compra. Gracias a su estructura visual, es posible rastrear f치cilmente la frecuencia y la proporci칩n de combinaciones espec칤ficas, lo que resulta beneficioso para el an치lisis de segmentaci칩n y la toma de decisiones estrat칠gicas en un contexto de negocios.
</p>


<p align="justify"> 游 Ejemplo:  
<br><br>
Se analizar치n los datos de preferencias de productos por regi칩n y segmento de cliente, para entender c칩mo los diferentes segmentos prefieren ciertos tipos de productos en distintas regiones.

```python
# Crear un DataFrame de ejemplo con datos de preferencias de productos
data = {
    'Regi칩n': ['Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur', 'Este', 'Oeste'],
    'Segmento_Cliente': ['Empresas', 'Empresas', 'Empresas', 'Empresas', 'Individuos', 'Individuos', 'Individuos', 'Individuos'],
    'Tipo_Producto': ['Electr칩nicos', 'Electr칩nicos', 'Muebles', 'Muebles', 'Electr칩nicos', 'Muebles', 'Electr칩nicos', 'Muebles'],
    'Preferencia_Compra': ['Alta', 'Baja', 'Media', 'Alta', 'Media', 'Alta', 'Baja', 'Media']
}

df = pd.DataFrame(data)
df
```

```python
# Crear el diagrama de categor칤as paralelas con Plotly
fig = px.parallel_categories(df,
                             dimensions=['Regi칩n', 'Segmento_Cliente', 'Tipo_Producto', 'Preferencia_Compra'],
                             color_continuous_scale=px.colors.sequential.Inferno,
                             title='Relaci칩n entre Regi칩n, Segmento de Cliente, Tipo de Producto y Preferencia de Compra')

# Mejorar el dise침o del gr치fico
fig.update_layout(
    title_font_size=18,
    template="plotly_white"  # Fondo blanco para mayor claridad
)

# Mostrar el gr치fico
fig.show()
```

En este diagrama de categor칤as paralelas:
<br><br>

- Cada eje representa una variable categ칩rica: Regi칩n, Segmento de Cliente, Tipo de Producto y Preferencia de Compra.
- Las l칤neas conectan las categor칤as entre los ejes, mostrando la proporci칩n y relaci칩n entre los valores de cada eje. Esto permite ver, por ejemplo, c칩mo el segmento de "Empresas" en la regi칩n "Norte" tiene una preferencia alta por productos electr칩nicos.
- Coloraci칩n: El color ayuda a resaltar las combinaciones con mayor frecuencia, facilitando la detecci칩n de patrones.

<p align="justify"><br>
Este gr치fico permite a una empresa analizar la relaci칩n entre ubicaci칩n geogr치fica, tipo de cliente, preferencia de producto y nivel de preferencia de compra. Esto es 칰til para tomar decisiones informadas sobre el inventario o campa침as de marketing dirigidas a ciertos segmentos y ubicaciones con mayor potencial de compra.








 # **<font color="DarkBlue">Generando un Violin Plots</font>**

<p align="justify">
Un <b>gr치fico de viol칤n</b> es una visualizaci칩n que combina caracter칤sticas de un box plot y un gr치fico de densidad. Este tipo de gr치fico representa la distribuci칩n de un conjunto de datos a trav칠s de su forma, mostrando la densidad de los datos a diferentes valores. En lugar de mostrar solo las estad칤sticas resumidas (como la mediana y los cuartiles) como en un box plot, el gr치fico de viol칤n proporciona una visi칩n m치s detallada de la distribuci칩n de los datos, lo que permite observar patrones, picos y variaciones en la densidad.

Los gr치ficos de viol칤n son 칰tiles para comparar la distribuci칩n de una variable entre diferentes grupos o categor칤as. Por ejemplo, en un contexto de negocios, se pueden utilizar para analizar la distribuci칩n de los ingresos por diferentes segmentos de mercado, permitiendo a las empresas identificar no solo el promedio de ingresos, sino tambi칠n c칩mo var칤an los ingresos dentro de cada segmento, ayudando en la segmentaci칩n de clientes y en la identificaci칩n de oportunidades de mercado.

<p align="justify"> 游 Ejemplo:

Se analizar치n los datos de preferencias de productos por regi칩n y segmento de cliente, para entender c칩mo los diferentes segmentos prefieren ciertos tipos de productos en distintas regiones.

```python
# Crear un DataFrame de ejemplo con datos de ingresos mensuales de clientes
np.random.seed(42)  # Para reproducibilidad de datos
data = {
    'Segmento_Cliente': ['Empresas'] * 500 + ['Individuos'] * 500,
    'Ingreso_Mensual': np.concatenate([
        np.random.normal(5000, 1000, 500),  # Ingresos para Empresas
        np.random.normal(3000, 800, 500)    # Ingresos para Individuos
    ])
}

df = pd.DataFrame(data)
df
```

```python
# Crear el gr치fico de viol칤n con Plotly
fig = px.violin(df,
                y='Ingreso_Mensual',
                x='Segmento_Cliente',
                box=True,  # Mostrar un box plot dentro del gr치fico de viol칤n
                points='all',  # Mostrar todos los puntos de datos
                title='Distribuci칩n de Ingresos Mensuales por Segmento de Cliente',
                labels={'Ingreso_Mensual': 'Ingreso Mensual', 'Segmento_Cliente': 'Segmento de Cliente'},
                color='Segmento_Cliente',  # Colorear por segmento
                template="plotly_white"  # Fondo blanco para mayor claridad
                )

# Mejorar el dise침o del gr치fico
fig.update_layout(
    title_font_size=18,
    yaxis_title="Ingreso Mensual",
    xaxis_title="Segmento de Cliente",
    yaxis_title_font_size=16,
    xaxis_title_font_size=16,
    violingap=0.5  # Espacio entre los gr치ficos de viol칤n
)

# Mostrar el gr치fico
fig.show()
```

En este gr치fico de viol칤n:
<br><br>

- **Ejes**: El eje X representa los segmentos de clientes (Empresas e Individuos), mientras que el eje Y muestra los ingresos mensuales.
- **Gr치ficos de viol칤n**: Cada gr치fico muestra la distribuci칩n de ingresos para cada segmento de clientes, permitiendo observar c칩mo se distribuyen los ingresos dentro de cada grupo.
- **Box plot**: Se incluye un box plot dentro del gr치fico de viol칤n, proporcionando informaci칩n adicional como la mediana y los cuartiles de los ingresos.
- **Puntos individuales**: Se muestran todos los puntos de ingresos individuales, lo que permite una comprensi칩n m치s profunda de la variabilidad de los datos.
