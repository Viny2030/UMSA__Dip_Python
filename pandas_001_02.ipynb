 # **<font color="DarkBlue">Tipos de operaciones</font>**

<p align="justify">
Las operaciones entre DataFrames y Series en Pandas permiten realizar c√°lculos y transformaciones eficientes sobre los datos. Estas operaciones son fundamentales para el an√°lisis de datos, ya que permiten manipular y combinar diferentes fuentes de datos, aplicar funciones matem√°ticas y l√≥gicas, y generar nuevas m√©tricas.
<br><br>
Ejemplo, una empresa est√° evaluando las ventas y costos de productos en diferentes tiendas, y aplica un descuento espec√≠fico a las ventas en funci√≥n de la tienda.




```python
import pandas as pd
```

```python
# Datos de ventas por tienda y producto
data = {
    'Tienda': ['Tienda A', 'Tienda A', 'Tienda B', 'Tienda B', 'Tienda C', 'Tienda C'],
    'Producto': ['Producto 1', 'Producto 2', 'Producto 1', 'Producto 2', 'Producto 1', 'Producto 2'],
    'Ventas': [20000, 15000, 18000, 16000, 22000, 19000],
    'Costo': [12000, 10000, 11000, 9500, 13000, 11500],
    'Cantidad Vendida': [500, 400, 450, 420, 550, 480]
}
```

```python
# Creando el DataFrame
df = pd.DataFrame(data)
```

```python
# Descuentos aplicados por tienda
descuentos = pd.Series({
    'Tienda A': 0.05,  # 5% de descuento
    'Tienda B': 0.10,  # 10% de descuento
    'Tienda C': 0.07   # 7% de descuento
})
```

 ## **<font color="DarkBlue">Operaciones Aritm√©ticas</font>**

<p align="justify">
Calcular la ganancia bruta (Ventas - Costo) para cada producto en cada tienda.

```python
df['Ganancia Bruta'] = df['Ventas'] - df['Costo']
```

```python
df[['Tienda', 'Producto', 'Ganancia Bruta']]
```

 ## **<font color="DarkBlue">Operaciones de Alineaci√≥n</font>**

<p align="justify">
Aplicar el descuento a las ventas para obtener las ventas ajustadas despu√©s del descuento.



```python
df['Ventas Ajustadas'] = df['Ventas'] * (1 - df['Tienda'].map(descuentos))
```

```python
df[['Tienda', 'Producto', 'Ventas Ajustadas']]
```

 ## **<font color="DarkBlue">Aplicaci√≥n de Funciones</font>**

<p align="justify">
Calcular el precio medio por unidad vendida en cada tienda.

```python
df['Precio Medio'] = df['Ventas'] / df['Cantidad Vendida']
```

```python
df[['Tienda', 'Producto', 'Precio Medio']]
```

 ## **<font color="DarkBlue">Operaciones entre DataFrames y Series</font>**

<p align="justify">
Para analizar c√≥mo los descuentos afectan la ganancia bruta, podemos calcular la ganancia neta ajustada.
<br><br>
Ejemplo: Calcular la ganancia neta ajustada aplicando los descuentos a las ventas y luego restar el costo.



```python
df['Ganancia Neta Ajustada'] = (df['Ventas'] * (1 - df['Tienda'].map(descuentos))) - df['Costo']
```

```python
df[['Tienda', 'Producto', 'Ganancia Neta Ajustada']]
```

 # **<font color="DarkBlue">Apply() y map()</font>**

<p align="justify">
En Pandas, aplicar y mapear funciones son m√©todos fundamentales para transformar y manipular datos en un DataFrame o en las Series.


 ## **<font color="DarkBlue">Aplicaci√≥n de Funciones (apply)</font>**

<p align="justify">
El m√©todo apply se utiliza para aplicar una funci√≥n a lo largo de un eje (filas o columnas) en un DataFrame o a cada elemento en una Serie. Es muy vers√°til y permite aplicar funciones personalizadas o integradas a los datos.
<br><br>
Uso en Series
<br><br>
Cuando se usa apply en una Serie, la funci√≥n se aplica a cada elemento de la Serie.
<br><br>
Uso en DataFrames
<br><br>
Cuando se usa apply en un DataFrame, se puede especificar el eje sobre el cual aplicar la funci√≥n (axis=0 para columnas y axis=1 para filas).



<p align="justify">
Supongamos que una empresa de comercio electr√≥nico tiene un DataFrame con informaci√≥n de ventas, incluyendo la fecha de venta, el monto total de la venta y la categor√≠a de producto. La empresa quiere calcular el descuento aplicado a cada venta basado en la categor√≠a del producto y agregar una columna que indique si una venta es alta, media o baja, bas√°ndose en el monto total de la venta.



```python
# Datos de ventas
data = {
    'Fecha': ['2024-01-15', '2024-02-20', '2024-03-05', '2024-04-10', '2024-05-22'],
    'Monto Total': [1500, 3000, 2500, 1000, 3500],
    'Categor√≠a': ['Electr√≥nica', 'Ropa', 'Electr√≥nica', 'Accesorios', 'Electr√≥nica']
}
```

```python
# Creaci√≥n del DataFrame
df = pd.DataFrame(data)
```

```python
# Descuentos por categor√≠a
descuentos = {
    'Electr√≥nica': 0.1,   # 10% de descuento
    'Ropa': 0.15,         # 15% de descuento
    'Accesorios': 0.05    # 5% de descuento
}
```

<p align="justify">
Calculando el Monto con Descuento
<br><br>
Primero, se aplicar√° una funci√≥n para calcular el monto ajustado despu√©s del descuento basado en la categor√≠a del producto.

```python
# Funci√≥n para aplicar el descuento basado en la categor√≠a
def aplicar_descuento(fila):
    descuento = descuentos[fila['Categor√≠a']]
    return fila['Monto Total'] * (1 - descuento)
```

```python
df['Monto con Descuento'] = df.apply(aplicar_descuento, axis=1)
df
```

<p align="justify">
Clasificando Ventas
<br><br>
Luego, se puede usar apply para clasificar las ventas como alta, media o baja basado en el monto total.

```python
# Funci√≥n para clasificar las ventas
def clasificar_venta(monto):
    if monto > 3000:
        return 'Alta'
    elif monto > 1500:
        return 'Media'
    else:
        return 'Baja'
```

```python
# Aplicar la funci√≥n a la columna 'Monto Total'
df['Clasificaci√≥n Venta'] = df['Monto Total'].apply(clasificar_venta)
df
```

 ## **<font color="DarkBlue">Mapeo de Funciones (map)</font>**

<p align="justify">
El m√©todo map se utiliza para aplicar una funci√≥n a cada elemento de una Serie. Es √∫til cuando se tiene una Serie y se quiere realizar una transformaci√≥n sencilla o se desea reemplazar valores con base en un mapeo.
<br><br>
Uso en Series
<br><br>
map es ideal para reemplazar valores en una Serie utilizando un diccionario de mapeo o una funci√≥n. Usando el mismo DataFrame, si la empresa quiere agregar una columna que muestre el descuento en formato porcentual basado en la categor√≠a del producto, se puede usar map.



```python
df['Descuento (%)'] = df['Categor√≠a'].map(lambda cat: descuentos[cat] * 100)
df
```

- apply: Permite aplicar funciones a lo largo de un eje de un DataFrame o a cada elemento de una Serie. Es √∫til para transformaciones m√°s complejas y para aplicar funciones personalizadas a cada fila o columna.

- map: Permite aplicar funciones o realizar reemplazos de valores en una Serie. Es ideal para transformaciones simples o mapeo directo de valores.

<p align="justify">
Otro ejemplo
<br><br>
Supongamos que una empresa de an√°lisis financiero tiene un DataFrame que contiene las previsiones de ingresos y costos para diferentes departamentos en distintas regiones. Los valores en el DataFrame est√°n en formato bruto, y la empresa quiere convertir todos los valores a miles de unidades para facilitar la lectura y comparaci√≥n.




```python
# Datos financieros
data = {
    'Regi√≥n': ['Norte', 'Sur', 'Este', 'Oeste'],
    'Ingresos Departamento A': [100000, 150000, 120000, 130000],
    'Ingresos Departamento B': [90000, 160000, 110000, 140000],
    'Costos Departamento A': [50000, 80000, 60000, 70000],
    'Costos Departamento B': [45000, 85000, 58000, 65000]
}
```

```python
# Creaci√≥n del DataFrame
df = pd.DataFrame(data)
df
```

<p align="justify">
La empresa quiere convertir todos los valores num√©ricos en el DataFrame de unidades completas a miles de unidades (es decir, dividir todos los valores por 1,000).



```python
# Funci√≥n para convertir valores a miles de unidades
def convertir_a_miles(x):
    return x / 1000
```

```python
# Aplicar la funci√≥n a cada elemento del DataFrame
df.iloc[:, 1:] = df.iloc[:, 1:].map(convertir_a_miles)
df
```

 # **<font color="DarkBlue">Ordenando y clasificando</font>**

<p align="justify">
Ordenar y clasificar son operaciones fundamentales en el an√°lisis de datos con Pandas. Estas operaciones permiten organizar los datos de manera que se pueda trabajar de manera m√°s eficiente y extraer informaci√≥n m√°s relevante.
<br><br>
A continuaci√≥n se explica c√≥mo funcionan estas operaciones, seguidas de ejemplos utilizando los m√©todos:

- sort_index(),
- sort_values() y
- rank().



 ## **<font color="DarkBlue">Ordenar por Valores</font>**

<p align="justify">
Ahora, la empresa quiere identificar qu√© producto ha tenido m√°s ventas en la regi√≥n "Norte" ordenando las ventas de mayor a menor.



```python
df
```

```python
# Ordenar el DataFrame por las ventas en la regi√≥n "Norte"
df_sorted_values = df.sort_values(by="Regi√≥n", ascending=False)
```

```python
# Mostrar el DataFrame ordenado por ventas en la regi√≥n Norte
df_sorted_values
```

 ## **<font color="DarkBlue">Ordenando por indice</font>**

<p align="justify">
Supongamos que la empresa quiere ordenar los datos por la regi√≥n de forma alfab√©tica.

```python
# Ordenar el DataFrame por el √≠ndice (Regi√≥n)
df_sorted_index = df_sorted_values.sort_index()
```

```python
# Mostrar el DataFrame ordenado por √≠ndice
df_sorted_index
```

 ## **<font color="DarkBlue">Clasificar con rank()</font>**

<p align="justify">
Finalmente, la empresa desea clasificar los productos en cada regi√≥n seg√∫n el n√∫mero de ventas, donde el producto m√°s vendido obtiene el rango m√°s alto.

```python
df
```

```python
# Clasificar los productos por sus ventas en cada regi√≥n
df_rank = df.iloc[:, 1:].rank(axis=1, method='dense', ascending=False)
df_rank
```

```python
# Agregar la columna 'Regi√≥n' de vuelta al DataFrame clasificado
df_rank.insert(0, 'Regi√≥n', df['Regi√≥n'])
df_rank
```
