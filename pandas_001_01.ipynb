 ## **<font color="DarkBlue">Indexaci贸n por etiquetas</font>**

<p align="justify">
<code>Pandas</code> proporciona un conjunto de m茅todos para lograr una indexaci贸n. Este es un protocolo estricto basado en la inclusi贸n. En el caso de indexaci贸n por etiquetas este es un protocolo estricto basado en la inclusi贸n. Cada etiqueta solicitada debe existir, tanto en el indice como en las columnas. Al seleccionar por ejemplo filas (indice) se incluye el l铆mite inicial indicado y el l铆mite final. <br><br>  Los n煤meros enteros son etiquetas v谩lidas, pero se refieren a las etiquetas del indice (cuando la etiqueta es num茅rica) y no a la posici贸n de cada fila. Lo mismo pasa cuando en el 铆ndice tenemos fechas.
</p>

~~~python
df.loc[fila inicial : fila final, columna inicial : columna final]
~~~


````python
# Datos de ventas
data = {
    'Producto': ['Producto A', 'Producto B', 'Producto C'],
    'Enero': [1000, 1500, 1200],
    'Febrero': [1100, 1400, 1300],
    'Marzo': [1200, 1600, 1250]
}
````

````python
# Creando el DataFrame
df = pd.DataFrame(data)
df.set_index('Producto', inplace=True)
df
````

````python
# Indexar por etiqueta
ventas_febrero = df.loc['Producto B', 'Febrero']
ventas_febrero
````

<p align="justify">
Vamos a seleccionar los ventas de Enero y de Febrero, es decir, las columnas <code>Enero</code> y <code>Febrero</code>.</p>



````python
df.loc[:, ["Enero", "Febrero"]] # [fila inicial : fila final , [lista de columnas]]
````

````python
df.loc[:,["Enero","Febrero"]].head(2)
````

<p align="justify">
O podemos seleccionar tambi茅n desde el producto B al producto C.</p>


````python
df.loc["Producto B":"Producto C",:] # [fila inicial : fila final , todas las columnas]
````

````python
df.loc["Producto B":"Producto C", ["Enero", "Marzo"]]
````

````python
df.loc[["Producto A"], ["Febrero", "Marzo"]]
````

 ## **<font color="DarkBlue">Indexaci贸n por posici贸n</font>**

<p align="justify">
<code>Pandas</code> proporciona un conjunto de m茅todos para obtener una indexaci贸n puramente basada en enteros. La sem谩ntica sigue de cerca lo definido en <code>Python</code> y <code>NumPy</code>. Al seleccionar, se incluye el l铆mite inicial $n$, mientras que se excluye el l铆mite final $m$, es decir desde $n$ hasta $m-1$ <br><br>  En los par谩metros de la selecci贸n, siempre se indica primero las filas y luego las columnas. En el caso de seleccionar de un DataFrame, se hace de la siguiente forma:
</p>

~~~python
df.iloc[fila inicial : fila final, columna inicial : columna final]
~~~


<p align="justify">
Vamos a seleccionar los valores de apertura y de cierre, es decir, las columnas <code>Open</code> y <code>Close</code>, pero usando este tipo de selecci贸n.</p>


````python
df.columns
````

````python
df.iloc[:,[0,2]].head(2)
````

<p align="justify">
O podemos seleccionar tambi茅n ...</p>


````python
df.iloc[0:3,:]
````

````python
df.iloc[[0],:]
````

<p align="justify">
Tambi茅n podemos seleccionar la posici贸n del dato.</p>


````python
df.iloc[0]
````

````python
df.iloc[[0]]
````

 ## **<font color="DarkBlue">Filtrado de Datos</font>**

<p align="justify">
<code>Pandas</code> nos permite llevar a cabo algunas opciones para filtrar nuestros datos, incluidas las m谩scaras booleanas y algunos m茅todos especiales. Con las m谩scaras booleanas, probamos nuestros datos contra alg煤n valor y obtenemos una estructura de la misma forma, excepto que est谩 llena de valores que son <code>True</code> o <code>False</code>.

````python
df
````

### **<font color="DarkBlue"> Mayor ```>``` Mayor igual ```>=``` Menor ```<``` Menor igual ```<=```</font>**

<p align="justify">
Filtramos los valores de Enero mayores a $1200$.</p>


````python
df[df.Enero > 1200]
````

<p align="justify">
Ahora filtramos los valores de Enero mayores a $1200$, pero con el m茅todo <code>query()</code>.</p>


````python
df.query("Enero > 1200")
````

### **<font color="DarkBlue">Dos o m谩s condiciones verdaderas ```&```</font>**

<p align="justify">
Para visualizar m谩s de dos condiciones verdaderas, es decir todas las condiciones verdaderas se usa el s铆mbolo &...</p>


````python
df
````

````python
df[(df.Enero > 1000) & (df.Febrero > 1100)]
````

````python
df.query("Enero > 1000 & Febrero > 1100")
````

````python
df.query("Enero > 1000 and Febrero > 1100")
````

### **<font color="DarkBlue">Dos o m谩s condiciones, alguna verdadera ```|```</font>**

<p align="justify">
Para visualizar m谩s de dos condiciones y alguna de ellas es verdadera, se usa el s铆mbolo |...</p>


````python
df
````

````python
df[(df.Febrero > 1400) | (df.Marzo > 1300)]
````

````python
df.query("Febrero > 1400 | Marzo > 1300")
````

````python
df.query("Febrero > 1400 or Marzo > 1300")
````

### **<font color="DarkBlue">Conjunto de valores ```between()```</font>**

<p align="justify">
Para un conjunto de valores entre un intervalo determinado...</p>


````python
df
````

````python
df[df.Febrero.between(1100, 1400)]
````

````python
df.query("Febrero >= 1100 & Febrero <= 1400")
````

````python
df.query("Febrero >= 1100 and Febrero <= 1400")
````

````python
df[df.Febrero.between(1100, 1400, inclusive="neither")] # excluye los extremos
````

````python
df.query("Febrero > 1100 & Febrero < 1400") # excluye los extremos
````
