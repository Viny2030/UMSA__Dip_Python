 ## **<font color="DarkBlue">Ausencia de datos</font>**

<p align="justify">
La ausencia de datos es un desaf칤o com칰n, y su correcta gesti칩n es crucial para obtener resultados fiables. Pandas facilita significativamente el trabajo con datos faltantes, haciendo que este proceso sea menos complejo y propenso a errores.
<br><br>
Por ejemplo, cuando se calcula estad칤sticas descriptivas como promedios, medianas o desviaciones est치ndar en Pandas, los valores faltantes se excluyen autom치ticamente del c치lculo, lo que te permite seguir obteniendo informaci칩n precisa sin interrumpir el flujo de trabajo.
<br><br>
En Pandas, se sigue una convenci칩n inspirada en el lenguaje de programaci칩n R, utilizando la notaci칩n <b>NA</b> (Not Available) para referirse a los datos faltantes. Estos valores pueden surgir de diversas situaciones:
<br><br>

- datos que nunca fueron capturados,
- errores en la recopilaci칩n o
- problemas t칠cnicos.

<p align="justify">
<br><br>
En aplicaciones empresariales, los datos faltantes pueden afectar decisiones clave. Por ejemplo, si en una base de datos de ventas faltan registros en ciertas fechas debido a fallos en el sistema, los an치lisis de rendimiento de ventas pueden resultar distorsionados, llevando a decisiones err칩neas sobre la demanda de productos.
<br><br>
Una buena pr치ctica al trabajar con datos faltantes es realizar un an치lisis exhaustivo de su patr칩n de ausencia.
<br><br>
Este an치lisis no solo ayuda a identificar problemas en el proceso de recopilaci칩n de datos, sino que tambi칠n puede revelar sesgos que podr칤an afectar los modelos predictivos o la interpretaci칩n de los resultados.
<br><br>
En un escenario de negocios, si los datos de clientes est치n incompletos, podr칤a indicar un sesgo en el p칰blico objetivo que la empresa est치 alcanzando, lo que puede influir negativamente en la planificaci칩n de campa침as de marketing o estrategias de ventas.
<br><br>
Pandas tambi칠n trata el valor especial de Python, <b>None</b>, como equivalente a <b>NA</b>. Esta unificaci칩n simplifica la limpieza y transformaci칩n de datos, permitiendo a los analistas concentrarse en resolver problemas empresariales complejos en lugar de gestionar la estructura de los datos.


游 Ejemplo:

```python
# Importar las bibliotecas necesarias
import pandas as pd
import numpy as np
```

```python
# Crear un DataFrame simulado con datos de ventas
data = {
    'Fecha': ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05'],
    'Region': ['Norte', 'Centro', 'Sur', 'Centro', 'Norte'],
    'Producto_A': [100, np.nan, 200, 150, np.nan],  # Faltan datos de ventas en d칤as espec칤ficos
    'Producto_B': [np.nan, 250, 180, np.nan, 300],  # Faltan datos en otras filas
    'Producto_C': [50, 60, np.nan, 80, 90]          # Faltan ventas de Producto C en un d칤a
}
```

```python
# Crear un DataFrame
df_ventas = pd.DataFrame(data)
```

```python
df_ventas
```

```python
# An치lisis b치sico de datos faltantes
print("\nResumen de datos faltantes:")
print(df_ventas.isna().sum())
```

 ### **<font color="DarkBlue">Estrategias de manejo de datos faltantes </font>**

```python
# Opci칩n 1: Eliminar filas con datos faltantes (dropna)
df_dropna = df_ventas.dropna()
print("\nDatos despu칠s de eliminar filas con valores faltantes:")
df_dropna
```

<p align="justify">
游 La funci칩n <code>dropna()</code> elimina cualquier fila que contenga valores faltantes, aunque esto puede llevar a la p칠rdida de informaci칩n valiosa.

```python
# Opci칩n 2: Rellenar los datos faltantes con la media (fillna)
df_fillna = df_ventas.fillna(df_ventas.mean(numeric_only=True))
print("\nDatos despu칠s de rellenar valores faltantes con la media:")
df_fillna
```

```python
# Opci칩n 3: Rellenar los datos faltantes con un valor espec칤fico (fillna con 0)
df_fillna_zeros = df_ventas.fillna(0)
print("\nDatos despu칠s de rellenar valores faltantes con ceros:")
df_fillna_zeros
```

 ### **<font color="DarkBlue">M칠todos de manejo de NA</font>**

<p align="justify">
  <strong>dropna</strong>: Este m칠todo filtra las etiquetas de los ejes (filas o columnas) en funci칩n de si los valores de cada etiqueta tienen datos faltantes. Puedes establecer umbrales variables para determinar cu치ntos datos faltantes se pueden tolerar antes de eliminar las etiquetas correspondientes.
<br><br>
  <strong>fillna</strong>: Este m칠todo completa los datos faltantes con un valor espec칤fico o utilizando un m칠todo de interpolaci칩n, como el relleno hacia adelante (<em>"ffill"</em>) o hacia atr치s (<em>"bfill"</em>), para imputar los valores ausentes.
<br><br>
  <strong>isna</strong>: Devuelve un conjunto de valores booleanos que indican qu칠 datos est치n faltantes o son <strong>NA</strong> (Not Available). Este m칠todo es 칰til para identificar los lugares donde los datos est치n incompletos.
<br><br>
  <strong>notna</strong>: Es la contraparte de <strong>isna</strong>. Retorna <strong>True</strong> para los valores que no son <strong>NA</strong> y <strong>False</strong> para los valores faltantes, lo que permite verificar la presencia de datos en el conjunto.


 ### **<font color="DarkBlue">Filtrado de datos faltantes</font>**

<p align="justify">
Existen algunas formas de filtrar los datos faltantes. Si bien siempre tiene la opci칩n de hacerlo manualmente mediante la indexaci칩n booleana <code>pandas.isna</code>, <code>dropna</code> puede resultar 칰til.
<br><br>
En una serie, el m칠todo <code>dropna</code> devuelve la serie solo con los datos no nulos y los valores del 칤ndice.



```python
df_ventas
```

```python
producto_A = df_ventas['Producto_A']
```

```python
producto_A.dropna()
```

Esto es lo mismo que hacer:

```python
producto_A[producto_A.notna()]
```

<p align="justify">
Ahora, para evitar que al eliminar los NA de un dataframe nos quedemos sin datos, ya que por defecto e eliminan todas las filas que contienen un valor faltante, entonces utilizamos el par치metro <code>how="all"</code> para eliminar solo las filas que sean todas NA.

```python
data = {
    'Fecha': [None, '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05'],
    'Region': [None, 'Centro', 'Sur', 'Centro', 'Norte'],
    'Producto_A': [None, np.nan, 200, 150, np.nan],
    'Producto_B': [np.nan, 250, 180, np.nan, 300],
    'Producto_C': [None, np.nan, np.nan, 80, np.nan]
}
```

```python
df_ventas = pd.DataFrame(data)
```

```python
df_ventas
```

```python
df_ventas.dropna()
```

```python
df_ventas.dropna(how="all")
```

Para eliminar columnas de la misma manera, utilizamos el par치metro <code>axis="columns"</code>



```python
df_ventas["Producto_D"] = None
```

```python
df_ventas
```

```python
df_ventas.info()
```

```python
df_ventas.dropna(axis="columns")
```

```python
df_ventas.dropna(axis="columns", how="all", inplace=True)
```

```python
df_ventas
```

<p align="Justify">
Supongamos que deseamos conservar 칰nicamente las filas que contengan como m치ximo una determinada cantidad de valores faltantes.
Podemos indicarlo con el par치metro <code>thresh</code>

```python
df_ventas
```

```python
df_ventas.dropna(thresh=2)
```

<p align="Justify">
Supongamos que deseamos eliminar 칰nicamente de la 칰ltima columna.
Podemos indicarlo con el par치metro <code>subset</code>

```python
df_ventas
```

```python
df_ventas.dropna(subset=["Producto_C"])
```

 ## **<font color="DarkBlue">Completando datos faltantes</font>**

<p align="justify">
En lugar de filtrar los datos faltantes (y posiblemente descartar otros datos junto con ellos), es posible rellenar los datos "vac칤os" de varias maneras...

```python
# Crear un DataFrame simulado con datos de env칤os
data = {
    'Fecha_Envio': ['2024-09-30', '2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04'],
    'Region': ['Norte', 'Centro', 'Sur', 'Centro', 'Norte'],
    'Tiempo_Entrega_dias': [3, np.nan, 5, np.nan, 2],  # Faltan tiempos de entrega para algunos env칤os
    'Costo_Envio': [100, 120, np.nan, 130, np.nan]     # Faltan costos de env칤o
}
```

```python
# Crear un DataFrame
df_envios = pd.DataFrame(data)
df_envios
```

 ### **<font color="DarkBlue">Estrategias de relleno de datos faltantes </font>**

```python
# 1. Rellenar los tiempos de entrega faltantes con la mediana de la columna

df_fillna_mediana = df_envios.copy()
df_fillna_mediana['Tiempo_Entrega_dias'] = df_fillna_mediana['Tiempo_Entrega_dias'].fillna(df_fillna_mediana['Tiempo_Entrega_dias'].median())
print("\nRellenar tiempos de entrega faltantes con la mediana:")
df_fillna_mediana
```

```python
# 2. Rellenar los costos de env칤o faltantes con un valor fijo (por ejemplo, 110)

df_fillna_fijo = df_envios.copy()
df_fillna_fijo['Costo_Envio'] = df_fillna_fijo['Costo_Envio'].fillna(110)
print("\nRellenar los costos de env칤o faltantes con un valor fijo (110):")
df_fillna_fijo
```
