 # **<font color="DarkBlue">Timestamp y Period</font>**

<p align="justify">
  En el an치lisis de datos, la representaci칩n y manipulaci칩n del tiempo son fundamentales, especialmente en 치reas como finanzas, log칤stica, y an치lisis de series temporales.
  <br><br>
  Pandas ofrece dos tipos principales de objetos para trabajar con datos relacionados con el tiempo: <code>Timestamp</code> y <code>Period</code>. Ambos juegan un papel crucial para modelar, analizar y procesar datos temporales de manera eficiente y precisa.
  <br><br>
  <strong>1. Timestamp:</strong> Este objeto representa un punto 칰nico en el tiempo. Es una generalizaci칩n del tipo <code>datetime</code> de Python, pero con funcionalidad ampliada para an치lisis de datos. Un <code>Timestamp</code> se define con:
  <br><br>


  $$T = Y, M, D, H, m, s, \mu, \text{tz}$$

  
  donde:
  <ul>
    <li>$Y$: A침o</li>
    <li>$M$: Mes</li>
    <li>$D$: D칤a</li>
    <li>$H$: Hora</li>
    <li>$m$: Minutos</li>
    <li>$s$: Segundos</li>
    <li>$\mu$: Microsegundos</li>
    <li>$\text{tz}$: Zona horaria (opcional)</li>
  </ul>
  <br>
  <p align="justify">
  Los <code>Timestamps</code> permiten realizar operaciones como comparaci칩n, diferencias temporales, y conversi칩n a diferentes formatos. Son ideales para indexar y manejar series temporales densas, donde cada punto de datos tiene una marca de tiempo 칰nica.
  <br><br>
  <strong>2. Period:</strong> A diferencia de <code>Timestamp</code>, un <code>Period</code> representa un intervalo fijo de tiempo, como un d칤a, un mes, o un trimestre. Su definici칩n incluye:
  <br><br>
  <center>
    $ P = (R, F) $
  </center>
  donde:
  <ul>
    <li>$R$: Punto de referencia temporal (por ejemplo, "2024-11")</li>
    <li>$F$: Frecuencia (como "M" para mensual o "Q" para trimestral)</li>
  </ul>
  <br>
  <p align="justify">
  Las operaciones con <code>Period</code> incluyen la agregaci칩n de datos en intervalos regulares, como calcular totales trimestrales o promedios mensuales. Adem치s, son 칰tiles para modelar y predecir tendencias en series temporales agrupadas por intervalos.
  <br><br>
  Tanto <code>Timestamp</code> como <code>Period</code> son fundamentales en pandas para trabajar con datos temporales y permiten un enfoque robusto para modelar tiempo en an치lisis de datos complejos.
</p>


```python
import pandas as pd
import numpy as np
import plotly.express as px
```

## **<font color="DarkBlue">Timestamp</font>**

游 Empezando a manipular objetos temporales

```python
# Crear un DataFrame con marcas de tiempo para ventas de un negocio
data = {
    "Producto": ["A", "B", "C"],
    "FechaVenta": ["2024-11-15 10:30:00", "2024-11-15 11:00:00", "2024-11-15 12:15:00"],
    "MontoVenta": [100, 200, 150]
}

data
```

```python
# Convertir la columna de fechas a Timestamp

df = pd.DataFrame(data)
df["FechaVenta"] = pd.to_datetime(df["FechaVenta"])
df
```

```python
# Calcular el tiempo transcurrido desde la 칰ltima venta

df["TiempoDesdeUltimaVenta"] = df["FechaVenta"].diff()
df
```

游 Generando un ejemplo con Timestamp

```python
# Generar datos de transacciones de una tienda con marcas de tiempo

np.random.seed(42)  # Asegura reproducibilidad
n_transacciones = 50
```

```python
# Generar un rango de fechas aleatorias para un mes

fechas = pd.date_range("2024-11-01", "2024-11-30", freq="h").to_pydatetime()
fechas_aleatorias = np.random.choice(fechas, size=n_transacciones, replace=True)
```

Generaci칩n de Fechas Aleatorias:
<br><br>
<p align="justify">
Utilizamos pd.date_range para crear una lista de fechas y horas que cubren todo noviembre.
Luego, seleccionamos al azar 50 marcas de tiempo utilizando np.random.choice.

```python
# Crear un DataFrame con productos, precios y marcas de tiempo

data = {
    "Producto": np.random.choice(["Producto A", "Producto B", "Producto C"], size=n_transacciones),
    "FechaVenta": fechas_aleatorias,
    "MontoVenta": np.random.randint(50, 500, size=n_transacciones)
}

df = pd.DataFrame(data)
df.head()
```

<p align="justify">
Creamos un conjunto de datos simulando transacciones, donde cada fila contiene un producto, una marca de tiempo, y un monto de venta.


```python
# Asegurarse de que las fechas est칠n ordenadas

df = df.sort_values("FechaVenta").reset_index(drop=True)
df.head()
```

<p align="justify">
Ordenamos las fechas para que el an치lisis sea cronol칩gico, algo fundamental en series temporales.


```python
# An치lisis con Timestamp:
# 1. Calcular el tiempo entre transacciones

df["TiempoDesdeUltimaVenta"] = df["FechaVenta"].diff()
df.head()
```

<p align="justify">
Usamos el m칠todo <code>diff()</code> en la columna FechaVenta para calcular el tiempo transcurrido desde la 칰ltima transacci칩n. Esto es 칰til para entender patrones de compra.


```python
# 2. Filtrar las ventas realizadas en un rango espec칤fico

inicio_rango = pd.Timestamp("2024-11-15 00:00:00")
fin_rango = pd.Timestamp("2024-11-16 23:59:59")
ventas_en_rango = df[(df["FechaVenta"] >= inicio_rango) & (df["FechaVenta"] <= fin_rango)]
ventas_en_rango.head()
```

<p align="justify">
Definimos un rango de fechas usando objetos Timestamp para seleccionar transacciones realizadas entre el 15 y el 16 de noviembre.


```python
# 3. Identificar el producto con mayor recaudaci칩n en todo el per칤odo

producto_top = df.groupby("Producto")["MontoVenta"].sum().idxmax()
producto_top
```

<p align="justify">
Agrupamos por Producto y sumamos las ventas para determinar cu치l gener칩 mayores ingresos.


```python
# Resultados

print("Datos procesados:\n", df)
print("\nVentas entre el 15 y el 16 de noviembre:\n", ventas_en_rango)
print("\nProducto con mayor recaudaci칩n:", producto_top)
```

## **<font color="DarkBlue">Period</font>**

```python
# Crear un DataFrame con datos trimestrales de ganancias
data = {
    "Trimestre": ["2024Q1", "2024Q2", "2024Q3", "2024Q4"],
    "Ganancia": [50000, 60000, 55000, 70000]
}

data
```

```python
# Convertir la columna de trimestres a Period

df = pd.DataFrame(data)
df["Trimestre"] = pd.PeriodIndex(df["Trimestre"], freq="Q")
df
```

```python
# Calcular la variaci칩n porcentual entre trimestres

df["VariacionPorcentual"] = df["Ganancia"].pct_change() * 100
df.VariacionPorcentual = df.VariacionPorcentual.round(2)
df
```

游 Generando un ejemplo con Period

```python
# Simulaci칩n de ventas trimestrales en un negocio

np.random.seed(42)  # Asegura reproducibilidad
n_periodos = 12  # 12 trimestres
```

```python
# Crear un rango de per칤odos trimestrales

periodos = pd.period_range(start="2020Q1", periods=n_periodos, freq="Q")
```

<p align="justify">
Creamos una serie de per칤odos utilizando pd.period_range con frecuencia trimestral (freq="Q"). Esto permite trabajar con intervalos de tiempo espec칤ficos como trimestres o a침os.


```python
# Generar un DataFrame con ingresos trimestrales y costos trimestrales

data = {
    "Periodo": periodos,
    "Ingresos": np.random.randint(5000, 15000, size=n_periodos),
    "Costos": np.random.randint(2000, 10000, size=n_periodos)
}

df = pd.DataFrame(data)
df
```

<p align="justify">
Asignamos ingresos y costos simulados a cada trimestre. Cada fila del DataFrame representa un per칤odo trimestral.


```python
# An치lisis con Period:
# 1. Calcular ganancias trimestrales

df["Ganancias"] = df["Ingresos"] - df["Costos"]
df
```

<p align="justify">
Restamos los costos de los ingresos para calcular las ganancias de cada trimestre, agregando esta informaci칩n como una nueva columna.


```python
# 2. Identificar el trimestre con mayores ganancias

max_ganancias_periodo = df.loc[df["Ganancias"].idxmax(), "Periodo"]
max_ganancias_periodo
```

<p align="justify">
Utilizamos <code>idxmax()</code> para localizar el 칤ndice del trimestre con mayores ganancias.


```python
# 3. Filtrar datos anuales

df["A침o"] = df["Periodo"].dt.year
ganancias_anuales = df.groupby("A침o")["Ganancias"].sum()
ganancias_anuales
```

<p align="justify">
Creamos una columna adicional con el a침o <code>df["Periodo"].dt.year</code> y agrupamos las ganancias por a침o para obtener un resumen anual.


```python
# 4. Conversi칩n a formato de fecha

df["Inicio_Periodo"] = df["Periodo"].dt.start_time
df["Fin_Periodo"] = df["Periodo"].dt.end_time
df
```

<p align="justify">
Usamos <code>dt.start_time</code> y <code>dt.end_time</code> para obtener las fechas de inicio y fin de cada trimestre, 칰til para an치lisis adicionales basados en fechas exactas.


```python
# Resultados

print("Datos procesados:\n", df)
print("\nTrimestre con mayores ganancias:", max_ganancias_periodo)
print("\nGanancias anuales:\n", ganancias_anuales)
```
