## **<font color="DarkBlue">GitHub</font>**

<p align="justify">
Para cargar datos desde GitHub en un cuaderno de Google Colab, puedes utilizar varias técnicas, dependiendo del formato de los datos y de su ubicación en el repositorio. Puedes acceder a el archivo directamente usando su URL cruda (raw URL).
<br><br>
<a href="https://github.com/cristiandarioortegayubro/BDS/tree/main/datasets">Enlace del GitHub</a>
<br><br>
A continuación se describe la cargar de un archivo desde GitHub usando su URL directa:

```python
import pandas as pd
url = "https://raw.githubusercontent.com/cristiandarioortegayubro/BDS/main/datasets/adult_census.csv"
df = pd.read_csv(url)
df.head()
```

<p align="justify">

## **<font color="DarkBlue">Google Drive</font>**

<p align="justify">
Cargar datos desde Google Drive a Google Colab es una práctica común y sencilla. Esto te permite acceder a archivos almacenados en tu Google Drive directamente desde el entorno de Colab.

<p align="justify">
Primero debes guardar el conjunto de datos en una carpeta de Google Drive. Para cargar estos datos en un entorno como Google Colab, puedes usar la biblioteca <code>google.colab</code> para montar tu unidad de Google Drive. Una vez montada, puedes acceder a los archivos en tu unidad como si fueran parte del sistema de archivos local.

```python
from google.colab import drive

# Montar Google Drive
drive.mount('/content/drive')
```

<p align="justify">
El código anterior te pedirá autorización para acceder a tu cuenta de Google y permitirá a Colab montar tu Google Drive en la ruta <code>/content/drive/</code>.
<br><br>
Una vez montado Google Drive, puedes acceder a los archivos dentro de tu unidad utilizando rutas estándar de Python. Por ejemplo, si tienes un archivo CSV en tu Google Drive, puedes cargarlo en un DataFrame de pandas de la siguiente manera:<code>/content/drive/My Drive/tu_archivo.csv</code>
<br><br>
Si el archivo está en una subcarpeta, asegúrate de especificar la ruta completa, por ejemplo:

```python
# Ruta al archivo en tu Google Drive
file_path = 'su ruta' # <-- debe indicar su ruta al archivo!!!

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(file_path)

# Mostrar las primeras filas del DataFrame
df.head()
```

Si prefieres cargar el archivo directamente, también puedes usar <code>gdown</code>. Esta es una herramienta útil para descargar archivos directamente desde Google Drive, utilizando solo el ID del archivo:
<BR><BR>
`https://drive.google.com/file/d/1XmyMKowOnrb6Bg9053XaHN-6yI5QEpXf/view?usp=sharing`
<BR><BR>
El ID del archivo es la parte después de `/d/` y antes de `/view?`. En este caso, sería `1XmyMKowOnrb6Bg9053XaHN-6yI5QEpXf`.
<BR><BR>
Luego se debe concatener la URL base de descarga (`https://drive.google.com/uc?id=`) de Google Drive con el ID del archivo. La estructura de la URL que debes construir es:
<BR><BR>
`https://drive.google.com/uc?id=1XmyMKowOnrb6Bg9053XaHN-6yI5QEpXf`

```python
ID = "1XmyMKowOnrb6Bg9053XaHN-6yI5QEpXf" # <-- debe indicar su ID
```

```python
import gdown

# URL base de descarga concatenado con el ID del archivo
url = f"https://drive.google.com/uc?id={ID}"

# Descargar el archivo
# quiet=True, desactiva la salida en pantalla durante la descarga (por defecto es False).
gdown.download(url, 'archivo.csv', quiet=True)

# Cargar el archivo descargado en un DataFrame
df = pd.read_csv('archivo.csv')
df.head()
```

## **<font color="DarkBlue">Repositorios</font>**

<p align="justify">
Es posible ingerir datos desde un repositorio web como el <a href = "https://archive.ics.uci.edu/">Repositorio de Aprendizaje Automático de UCI</a>, que contiene una amplia colección de conjuntos de datos utilizados para investigaciones y educación en Machine Learning. El siguiente código lo encontrará en este <a href="https://archive.ics.uci.edu/dataset/2/adult">link</a>.

```python
!pip install ucimlrepo
from ucimlrepo import fetch_ucirepo

# Obtén el dataset con id=2 (Adult dataset)
adult = fetch_ucirepo(id=2)

# Los datos (características) ya están en un DataFrame de pandas
X = adult.data.features
y = adult.data.targets

# Combina las variables explicativas y la objetivo en un solo DataFrame
df = pd.concat([X, y], axis=1)
df.head()
```

<p align="justify">
También es posbile cargar el conjunto de datos usando <code>fetch_openml</code> de <code>scikit-learn</code>.
<br><br>
<a href="https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_openml.html">Documentación</a>

```python
from sklearn.datasets import fetch_openml

# Cargar el conjunto de datos Adult Census desde sklearn
adult_census = fetch_openml(name="adult", version=2, as_frame=True)

# El conjunto de datos se carga como un Bunch, de donde podemos extraer el DataFrame
df = adult_census.frame
df.head()
```

```python
# Cargar el conjunto de datos Adult Census desde sklearn versión 3
adult_census = fetch_openml(name="adult", version=3, as_frame=True)

# El conjunto de datos se carga como un Bunch, de donde podemos extraer el DataFrame
df = adult_census.frame
df.head()
```

## **<font color="DarkBlue">Unidad local</font>**

Para cargar datos desde tu máquina local en Google Colab y trabajar con ellos sin necesidad de almacenar los datos en un repositorio en línea.

```python
from google.colab import files

# Subir el archivo
uploaded = files.upload()
```

Esto abrirá un cuadro de diálogo que te permitirá seleccionar y subir archivos desde tu undidad local. Después de subir el archivo, se almacenará en el entorno de Colab. Puedes usar pandas para leer archivos CSV:

```python
# Hemos subido un archivo llamado 'adult_census.csv'
filename = list(uploaded.keys())[0]  # Obtén el nombre del archivo subido
df = pd.read_csv(filename)
```

Si necesitas cargar otros tipos de archivos (como Excel, JSON, etc.), puedes usar las funciones adecuadas de pandas (como` pd.read_excel`, `pd.read_json`, etc.).
